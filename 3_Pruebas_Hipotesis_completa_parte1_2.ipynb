## Â¿QuÃ© es una prueba de hipÃ³tesis?

Una **prueba de hipÃ³tesis** es un procedimiento estadÃ­stico utilizado para tomar decisiones o llegar a conclusiones acerca de una **poblaciÃ³n**, basÃ¡ndose en datos extraÃ­dos de una **muestra**.

## Â¿CuÃ¡les son las bases de las hipÃ³tesis?

### HipÃ³tesis nula (Hâ‚€)
La **hipÃ³tesis nula** es una afirmaciÃ³n que representa el estado de equilibrio en un experimento o estudio. Por ejemplo, es la suposiciÃ³n de que no hay diferencia significativa entre grupos.

Se establece como punto de partida para la inferencia estadÃ­stica, y solo se rechaza si la evidencia de los datos es lo suficientemente fuerte.

**Ejemplo:**

### HipÃ³tesis alternativa (Hâ‚)
La **hipÃ³tesis alternativa** contradice a la hipÃ³tesis nula. Postula que sÃ­ existe una diferencia significativa.

Si los datos aportan suficiente evidencia en contra de Hâ‚€, se rechaza Hâ‚€ como la explicaciÃ³n mÃ¡s plausible.

En resumen, el proceso de prueba de hipÃ³tesis consiste en:

## Â¿CuÃ¡les son los parÃ¡metros importantes de dichas pruebas?

## 1.2. Errores Tipo I y II en pruebas de hipÃ³tesis
Estos errores son inherentes a las decisiones estadÃ­sticas. Tenemos dos distribuciones (Hâ‚€ y Hâ‚) y un **umbral** o **regiÃ³n crÃ­tica** para **rechazar Hâ‚€**.
1. **Error Tipo I (Î±):** Rechazamos Hâ‚€ cuando en realidad es verdadera. Es un **falso positivo**.
- **Control**: se limita fijando el **nivel de significancia** Î± (si *p* â‰¤ Î±, se rechaza Hâ‚€).
2. **Error Tipo II (Î²):** No rechazamos Hâ‚€ cuando en realidad es falsa. Hâ‚ verdadera. Es un **falso negativo**.
- **Potencia de una prueba**: $1 - \beta$ = probabilidad de **detectar un efecto real** (rechazar Hâ‚€ cuando Hâ‚ es verdadera).

### Al bajar **Î±** suele subir **Î²** (y viceversa).
Controlar el nivel de significancia (Î±) nos ayuda a limitar la probabilidad de cometer un error Tipo I, pero aumenta el error Tipo II.

### Â¿**CÃ³mo reducir ambos** (sin solo mover el umbral)?
 **Mayor tamaÃ±o de muestra**, **menor variabilidad** (mejor diseÃ±o/mediciÃ³n, menos ruido), **efectos mÃ¡s grandes (mayor separaciÃ³n entre grupos Hâ‚€ y Hâ‚** o **pruebas mÃ¡s potentes**.

## 2.1. Prueba de normalidad de Shapiroâ€“Wilk

La prueba de **Shapiroâ€“Wilk** evalÃºa si una muestra proviene de una distribuciÃ³n normal. Es especialmente Ãºtil cuando se trabaja con muestras pequeÃ±as o moderadas (n < 50 o incluso hasta 2000 segÃºn algunas implementaciones).

### HipÃ³tesis:

- **HipÃ³tesis nula (Hâ‚€):** los datos provienen de una distribuciÃ³n normal.
- **HipÃ³tesis alternativa (Hâ‚):** los datos no provienen de una distribuciÃ³n normal.

El estadÃ­stico de la prueba se denota por \( W \) y se calcula mediante la siguiente fÃ³rmula:

$W = \frac{\left( \sum_{i=1}^{n} a_i x_{(i)} \right)^2}{\sum_{i=1}^{n} (x_i - \bar{x})^2}$

donde:

- $ x_{(i)} $ son los valores ordenados de menor a mayor en la muestra (orden estadÃ­stico).
- $ \bar{x} $ es la media de la muestra.
- $ a_i$ son coeficientes calculados a partir de la matriz de covarianza esperada de los ordenamientos bajo normalidad.
- $ n $ es el tamaÃ±o de la muestra.

Este estadÃ­stico compara la varianza explicada por el modelo de normalidad con la varianza total de la muestra. Cuanto mÃ¡s se acerque $ W$ a 1, mÃ¡s compatibles son los datos con una distribuciÃ³n normal.

---

#### DecisiÃ³n basada en valor-p:

- Si $ p > \alpha $, **no se rechaza Hâ‚€**
- Si $ p \leq \alpha $, **se rechaza Hâ‚€**

> En la prÃ¡ctica, se utiliza el valor-p que proporciona el software (por ejemplo, `scipy.stats.shapiro`) para la decisiÃ³n, ya que los coeficientes $ a_i $ no se calculan manualmente.


#### Â¿CÃ³mo se calculan los coeficientes $ a_i $?

Los coeficientes $ a_i $ dependen del tamaÃ±o de la muestra $ n $, y se obtienen mediante los siguientes pasos:

1. **Se calcula el vector $ m $**:  
   Es el vector de medias esperadas de los ordenamientos de una muestra aleatoria de tamaÃ±o $n$ de una distribuciÃ³n normal estÃ¡ndar $ N(0,1) $.

2. **Se calcula la matriz de covarianza $ V $** asociada a estos ordenamientos.

3. **Los coeficientes $a$** se obtienen resolviendo la siguiente ecuaciÃ³n de normalizaciÃ³n:

  $   a = \frac{m^T V^{-1}}{\sqrt{m^T V^{-1} V^{-1} m}} $

> Esta operaciÃ³n asegura que los coeficientes $a_i$ estÃ¡n optimizados para maximizar la potencia de la prueba bajo la hipÃ³tesis alternativa.

---
#### **Nota**: Estos cÃ¡lculos son complejos y no se realizan manualmente. Las librerÃ­as estadÃ­sticas como `scipy.stats.shapiro()` ya incluyen estos valores precomputados para distintos tamaÃ±os muestrales.
---
> **CÃ¡lculo de los coeficientes \( a_i \):**  
>  
> Los coeficientes $ a_i $ del estadÃ­stico de Shapiroâ€“Wilk se derivan de la **matriz de covarianza esperada** de los ordenamientos bajo normalidad.  
>  
> Este cÃ¡lculo puede ser **computacionalmente costoso**, ya que suele  **calcularse la inversa de  matrices**, por ejemplo usando **determinantes**.  
>  
> Por esa razÃ³n, para cada tamaÃ±o muestral $n$, estos coeficientes suelen estar **precalculados** e incluidos en bibliotecas como `scipy`.---

### 2.1.1. ðŸ“˜Ejemplo de Shapiro-Wilk

````python
from scipy.stats import shapiro
import seaborn as sns
import matplotlib.pyplot as plt

# Datos de ejemplo
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]

# Aplicar la prueba de Shapiro-Wilk
stat, p = shapiro(data)
print('EstadÃ­stico W = %.3f, valor-p = %.3f' % (stat, p))

# InterpretaciÃ³n del valor-p
alpha = 0.05
if p > alpha:
    print("p > 0.05: No se rechaza H0: probablemente los datos provienen de una distribuciÃ³n normal.")
else:
    print("p <= 0.05: Se rechaza H0: probablemente los datos NO provienen de una distribuciÃ³n normal.")

# VisualizaciÃ³n de la distribuciÃ³n
sns.histplot(data, kde=True)
plt.title("DistribuciÃ³n de la muestra")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
````

````python
# Generando nÃºmeros aleatorios con distribuciÃ³n normal
import numpy as np
import seaborn as sns
from scipy.stats import shapiro

np.random.seed(1984)  # Semilla para reproducibilidad

# GeneraciÃ³n de muestra
muestras = np.random.normal(loc=0, scale=1, size=50)

# Aplicando la prueba de Shapiro-Wilk
stat, p = shapiro(muestras)

print(f'EstadÃ­stico W = {stat:.3f}, p-value = {p:.3f}')

# Nivel de significancia
alpha = 0.05
if p > alpha:
    print('Si (p > alpha): La muestra parece provenir de una distribuciÃ³n normal (no se rechaza H0)')
else:
    print('Si (p <= alpha)La muestra no parece provenir de una distribuciÃ³n normal (se rechaza H0)')

# VisualizaciÃ³n de la distribuciÃ³n
sns.histplot(muestras, kde=True)
````

##3.1. Prueba t de Student

Una **prueba t** es una herramienta de estadÃ­stica inferencial utilizada para determinar si existe una **diferencia significativa entre medias**.

Se aplica cuando:

- Los datos provienen de una distribuciÃ³n aproximadamente normal
- El tamaÃ±o de la muestra es pequeÃ±o (generalmente n < 30)
---
Existren tres modalidades de la prueba t, se utilizan para comparar:

- Una muestra con un valor fijo (media hipotÃ©tica).
- Dos muestras independientes entre sÃ­.
- Una misma muestra, antes y despuÃ©s de una intervenciÃ³n (muestras pareadas).

### 3.1.1. FormulaciÃ³n de hipÃ³tesis **dos colas vs. una cola** en pruebas t-Student:
| Tipo de prueba t           | Hâ‚€ (dos colas)             | Hâ‚ (dos colas)                 | Hâ‚ (una cola)                      |
|----------------------------|----------------------------|--------------------------------|------------------------------------|
| Una muestra                | Î¼ = Î¼â‚€                     | Î¼ â‰  Î¼â‚€                         | Î¼ > Î¼â‚€  **o**  Î¼ < Î¼â‚€              |
| Muestras independientes    | Î¼â‚ = Î¼â‚‚                    | Î¼â‚ â‰  Î¼â‚‚                        | Î¼â‚ > Î¼â‚‚  **o**  Î¼â‚ < Î¼â‚‚            |

> â„¹ï¸ **Nota importante:**  
> Si lo que se desea comparar es la **varianza** entre dos grupos (es decir, su dispersiÃ³n), se utiliza la **prueba F de Fisher**, no la prueba t.



### 3.1.2. Prueba t **de una muestra**

Esta prueba permite evaluar si la **media de una muestra** difiere significativamente de un valor poblacional conocido o hipotÃ©tico.

**HipÃ³tesis:**

- Hâ‚€: Î¼ = Î¼â‚€ â†’ la media de la muestra es igual al valor hipotÃ©tico
- Hâ‚: Î¼ â‰  Î¼â‚€ â†’ la media de la muestra es distinta

**Supuestos:**

- Los datos deben ser aproximadamente normales
- La varianza poblacional es desconocida
- Los datos son independientes
---
**EstadÃ­stico:**

$ t = \dfrac{\bar{x} - \mu_0}{s / \sqrt{n}} $

Donde:  
- $\bar{x}$: media muestral  
- $\mu_0$: valor hipotÃ©tico de la media  
- $s$: desviaciÃ³n estÃ¡ndar muestral  
- $n$: tamaÃ±o de la muestra


**Ejemplo:**

Evaluar si el puntaje promedio de un grupo de estudiantes **difiere de una calificaciÃ³n mÃ­nima esperada** (por ejemplo, Î¼â‚€ = 70).


####  ðŸ“˜**Ejemplo: Prueba t de una muestra**


````python
# Ejemplo de prueba t de una muestra
import numpy as np
from scipy.stats import ttest_1samp

# Semilla para reproducibilidad
np.random.seed(1984)

# Generamos una muestra aleatoria de edades (enteros entre 0 y 49)
ages = np.random.randint(0, 50, size=10)
print("Muestra de edades:", ages)

# Media muestral
ages_mean = np.mean(ages)
print("Media muestral:", round(ages_mean, 2))

# HipÃ³tesis:
# H0: Î¼ = 30 (la media poblacional es igual a 30)
# H1: Î¼ â‰  30 (la media poblacional es diferente de 30)

# Realizando la prueba t de una muestra
t_stat, p_value = ttest_1samp(ages, popmean=30)

print(f"\nEstadÃ­stico t = {t_stat:.4f}")
print(f"Valor-p = {p_value:.4f}")

# Nivel de significancia
alpha = 0.05

if p_value < alpha:
    print("Se rechaza la hipÃ³tesis nula (H0). Eso significa que la media poblacional NO es igual a 30")
else:
    print("No se rechaza la hipÃ³tesis nula (H0). Eso significa que la media poblacional es igual a 30")
````

### 3.1.3.  Prueba t de **dos muestras (independientes)**

La **prueba t de dos muestras independientes** se utiliza para comparar las medias de **dos grupos diferentes y no relacionados** entre sÃ­.

Sirve para evaluar si existe **evidencia estadÃ­stica** suficiente de que sus medias poblacionales difieren de forma significativa.


** Supuestos:**
- Grupos independientes entre sÃ­
- Datos aproximadamente normales
- Varianzas iguales (aunque se puede ajustar si no lo son)

**HipÃ³tesis:**
- Hâ‚€: Î¼â‚ = Î¼â‚‚ â†’ no hay diferencia entre las medias
- Hâ‚: Î¼â‚ â‰  Î¼â‚‚ â†’ sÃ­ hay diferencia entre las medias
---
 Esta prueba t se debe aplicar despuÃ©s de haber aplicado la prueba F de Fisher para comparar varianzas.

Solo en la prueba t para muestras independientes es relevante distinguir entre varianzas iguales o diferentes.

#### 3.1.3.1.  **Prueba t con varianzas iguales**

El estadÃ­stico de la prueba se calcula con la fÃ³rmula:

$$
t = \frac{|\bar{x} - \bar{y}|}{s \cdot \sqrt{\frac{1}{n_x} + \frac{1}{n_y}}}
$$

Donde:

- $\bar{x}, \bar{y}$: medias muestrales  
- $n_x, n_y$: tamaÃ±os de muestra  
- $s$: desviaciÃ³n estÃ¡ndar combinada (pooled)

La desviaciÃ³n estÃ¡ndar combinada se calcula como:

$$
s = \sqrt{ \frac{(n_x - 1)s_x^2 + (n_y - 1)s_y^2}{n_x + n_y - 2} }
$$

**Grados de libertad:**

$$
\nu = n_x + n_y - 2
$$


#### 3.1.3.2. **Prueba t con varianzas distintas**

El estadÃ­stico de la prueba se calcula con la fÃ³rmula:

$$
t = \frac{|\bar{x} - \bar{y}|}{\sqrt{ \frac{s_x^2}{n_x} + \frac{s_y^2}{n_y} }}
$$

Donde:

- $s_x^2, s_y^2$: varianzas muestrales  
- $n_x, n_y$: tamaÃ±os de muestra

**Grados de libertad aproximados (fÃ³rmula de Welch):**

$$
\nu = \frac{ \left( \frac{s_x^2}{n_x} + \frac{s_y^2}{n_y} \right)^2 }
{ \frac{ \left( \frac{s_x^2}{n_x} \right)^2 }{n_x - 1} +
  \frac{ \left( \frac{s_y^2}{n_y} \right)^2 }{n_y - 1} }
$$


````python
import numpy as np
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt

# Para reproducibilidad
np.random.seed(1984)

# Simulamos dos muestras independientes
data1 = np.random.randint(0, 50, size=100)          # muestra 1: enteros uniformes [0, 49]
data2 = np.random.normal(loc=20, scale=10, size=100) # muestra 2: normal con media=20, desv=10

# EstadÃ­sticas descriptivas
print(f"Media data1: {np.mean(data1):.2f}")
print(f"Media data2: {np.mean(data2):.2f}")
print(f"Desv. estÃ¡ndar data1: {np.std(data1, ddof=1):.2f}")
print(f"Desv. estÃ¡ndar data2: {np.std(data2, ddof=1):.2f}")

# Prueba t de dos muestras independientes
t_stat, p_value = ttest_ind(data1, data2, equal_var=True) #si las varianzas son muy diferentes, puede no cumplirse el supuesto de homogeneidad, se recomienda usar equal_var=False en ttest_ind().

print(f"\nEstadÃ­stico t: {t_stat:.4f}")
print(f"Valor-p: {p_value:.4f}")

# DecisiÃ³n con Î± = 0.05
alpha = 0.05
if p_value < alpha:
    print("Se rechaza la hipÃ³tesis nula (Hâ‚€): las medias son significativamente diferentes (medias NO son iguales).")
else:
    print("No se rechaza la hipÃ³tesis nula (Hâ‚€): no hay evidencia suficiente de diferencia (medias son iguales).")

# VisualizaciÃ³n de las distribuciones
sns.histplot(data1, kde=True, label="Grupo 1", stat="density")
sns.histplot(data2, kde=True, label="Grupo 2", stat="density")
plt.legend()
plt.title("Distribuciones de los dos grupos")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
````

- Si valor-p â‰¥ Î± â†’ **no se rechaza Hâ‚€**

  No existe diferencia significativa entre los resultados obtenidos por los dos mÃ©todos.

### ðŸ“Œ RelaciÃ³n entre prueba t a dos colas y una cola

- **Dos colas (two-sided)**  
  Hâ‚€: Î¼d = 0  
  Hâ‚: Î¼d â‰  0  
  â†’ p-value dos colas (por defecto en Python)

- **Una cola (one-sided)**  
  Hâ‚€: Î¼d = 0  
  Hâ‚: Î¼d < 0 Ã³ Î¼d > 0  
  â†’ p-value una cola = p-value dos colas / 2 (si t va en la direcciÃ³n de Hâ‚)

####ðŸ“˜ Este ejemplo en python

````python
import numpy as np
import scipy.stats as stats

# ---------------------------
# Datos: pesos antes y despuÃ©s
# ---------------------------
antes =   np.array([124,157, 98,190,103,135,149,176,200,180,256])
despues = np.array([142,157, 96,212,116,134,150,184,209,180,269])

# Diferencias
d = antes - despues
n = len(d)
d_mean = np.mean(d)
d_std = np.std(d, ddof=1)

print("Resumen de diferencias:")
print(f"n = {n}")
print(f"Media de d = {d_mean:.3f}")
print(f"DesviaciÃ³n estÃ¡ndar de d = {d_std:.3f}")
print()

# ---------------------------
# Prueba t pareada
# ---------------------------
t_stat, p_value_two_sided = stats.ttest_rel(antes, despues)

print("Prueba t pareada (dos colas):")
print(f"EstadÃ­stico t = {t_stat:.4f}")
print(f"p-value (dos colas) = {p_value_two_sided:.6f}")

# ---------------------------
# Para una cola (hipÃ³tesis: aumento de peso â†’ diferencia negativa antes - despuÃ©s)
# ---------------------------
p_value_one_sided = p_value_two_sided / 2

print("\nPrueba t pareada (una cola, H1: peso despuÃ©s > peso antes):")
print(f"EstadÃ­stico t = {t_stat:.4f}")
print(f"p-value (una cola) = {p_value_one_sided:.6f}")

# DecisiÃ³n al 99% de confianza
alpha = 0.01
if p_value_one_sided <= alpha:
    print(f"â†’ Con Î±={alpha}, se rechaza H0: aumento significativo de peso")
else:
    print(f"â†’ Con Î±={alpha}, no se rechaza H0: no hay evidencia suficiente de aumento significativo")
````

##3.2. Prueba F de Fisher â€” ComparaciÃ³n de varianzas

La prueba F de Fisher se utiliza para **comparar las varianzas de dos muestras** y verificar si son estadÃ­sticamente diferentes.  
Es un paso previo importante antes de aplicar la prueba t de Student para muestras independientes, ya que determina si se deben asumir varianzas iguales o diferentes.

---

**EstadÃ­stico de la prueba**

Para que el valor de F sea **mayor o igual a 1**, se recomienda colocar **la muestra con mayor varianza en el numerador**:

$$
F = \frac{s_x^2}{s_y^2} \quad \text{donde } F \geq 1
$$

Si al calcular F se obtiene un valor **menor que 1**, se deben **invertir los tÃ©rminos**:

$$
F = \frac{s_y^2}{s_x^2} \quad \text{donde } F \geq 1
$$

---
**Ejemplo:**

- $s_x^2 = 0.98$, $s_y^2 = 0.45$  
- $F = \dfrac{0.98}{0.45} = 2.178$

Grados de libertad:

- $v_1 = n_x - 1 = 20 - 1 = 19$  
- $v_2 = n_y - 1 = 8 - 1 = 7$

---

**ComparaciÃ³n con el valor crÃ­tico**

Supongamos un nivel de significancia del 5%:

- Valor crÃ­tico (de tabla): $F_{crÃ­tico} = 3.454$

Comparamos:

$$
2.178 \; < \; 3.454
$$

---
**ConclusiÃ³n:**

Como el valor calculado de F **no supera** el valor crÃ­tico:

> **No se rechaza la hipÃ³tesis nula $H_0$**  
> Es decir, **se asume que las varianzas son iguales**

--> Siguiente paso: aplicar la prueba t de Student **con varianzas iguales**.


````python
import scipy.stats as stats

# ---------------------------
# Datos de ejemplo
# ---------------------------
s2_x = 0.98   # Varianza muestra X
s2_y = 0.45   # Varianza muestra Y
n_x  = 20     # TamaÃ±o muestra X
n_y  = 8      # TamaÃ±o muestra Y
alpha = 0.05  # Nivel de significancia (5%)

# ---------------------------
# EstadÃ­stico F
# Siempre poner la mayor varianza en el numerador
# ---------------------------
if s2_x >= s2_y:
    F = s2_x / s2_y
    v1, v2 = n_x - 1, n_y - 1
else:
    F = s2_y / s2_x
    v1, v2 = n_y - 1, n_x - 1

# ---------------------------
# Valor crÃ­tico de la tabla F
# Percent point function (cuantil) con 1-alpha
# ---------------------------
F_critico = stats.f.ppf(1 - alpha, v1, v2)

# ---------------------------
# Resultado
# ---------------------------
print(f"EstadÃ­stico F = {F:.4f}")
print(f"Grados de libertad: v1={v1}, v2={v2}")
print(f"Valor crÃ­tico F (Î±={alpha*100:.1f}%): {F_critico:.4f}")

if F > F_critico:
    print("Se rechaza H0: Por lo tanto, las varianzas son diferentes.")
else:
    print("No se rechaza H0: Por lo tanto, las varianzas son iguales.")
````

#### Este mismo ejemplo con decisiÃ³n basada en p-value
Ahora en lugar de comparar con el valor crÃ­tico de F, tambiÃ©n podemos usar el p-value de la prueba F. Esto da un criterio mÃ¡s intuitivo:
 >Recordemos que p-value es la probabilidad, asumiendo **Hâ‚€** verdadera, de observar un estadÃ­stico **F** **tan grande o mayor** que el obtenido.

```# CÃ¡lculo del p-value
p_value = 1 - stats.f.cdf(F, v1, v2)
```
 - `F`: F calculado.
  - `v1`, `v2`: grados de libertad (numerador y denominador).  
*(Equivalente numÃ©ricamente: `stats.f.sf(F, v1, v2)`.)*
---
Si p_value es pequeÃ±o (â‰¤ Î±), hay evidencia para **rechazar Hâ‚€**:
* Si p â‰¤ Î±, se rechaza Hâ‚€: las varianzas son diferentes.
* Si p > Î±, no se rechaza Hâ‚€: se asumen varianzas iguales.

---


````python
# ---------------------------
# CÃ¡lculo del p-value
# DistribuciÃ³n F es asimÃ©trica â†’ usamos cola superior
# ---------------------------
p_value = 1 - stats.f.cdf(F, v1, v2)

# ---------------------------
# Resultados
# ---------------------------
print(f"EstadÃ­stico F = {F:.4f}")
print(f"Grados de libertad: v1={v1}, v2={v2}")
print(f"p-value = {p_value:.6f}")

if p_value <= alpha:
    print(f"p-value â‰¤ {alpha} â†’ Se rechaza H0: Las varianzas son diferentes.")
else:
    print(f"p-value > {alpha} â†’ No se rechaza H0: Se asumen varianzas iguales.")
````

ConclusiÃ³n: El p-value = 0.148293 > 0.05, asÃ­ que no rechazamos Hâ‚€ y seguimos con la prueba t de Student con varianzas iguales.

###3.2.1 Explicacion avanzada del cÃ¡lculo de p-value para una cola vs. dos colas

### Una cola (cola superior)
Si $H_1: \sigma_A^2 > \sigma_B^2$.

```python
# CÃ¡lculo del p-value (cola superior)
p_value = 1 - stats.f.cdf(F, v1, v2)   
# equivalente a:
p_value = stats.f.sf(F, v1, v2)


#### CÃ¡lculo de p-values en Python para prueba F

````python
# F-test de varianzas con SciPy â€” devuelve SOLO el p-value
# Supone normalidad e independencia en ambos grupos.

import numpy as np
from scipy import stats

def ftest_var_pvalue(x, y, alternative="two-sided"):
    """
    Prueba F para comparar varianzas (s_x^2 / s_y^2).
    alternative: 'greater' (H1: Ïƒ_x^2 > Ïƒ_y^2), 'less' (H1: Ïƒ_x^2 < Ïƒ_y^2), 'two-sided' (H1: Ïƒ_x^2 â‰  Ïƒ_y^2)
    Devuelve: p-value (float)
    """
    x = np.asarray(x, float)
    y = np.asarray(y, float)
    n1, n2 = x.size, y.size
    if n1 < 2 or n2 < 2:
        raise ValueError("Cada grupo debe tener â‰¥ 2 observaciones.")

    s1 = x.var(ddof=1)
    s2 = y.var(ddof=1)
    F = s1 / s2
    v1, v2 = n1 - 1, n2 - 1

    if alternative == "greater":          # cola superior
        print("pvalue greater cola superior:")
        p = stats.f.sf(F, v1, v2)
    elif alternative == "less":           # cola inferior
        print("pvalue less cola inferior:")
        p = stats.f.cdf(F, v1, v2)
    elif alternative == "two-sided":      # dos colas: F no es simÃ©trica
        print("pvalue two-sided:")
        p = 2 * min(stats.f.cdf(F, v1, v2), stats.f.sf(F, v1, v2))
        p = min(1.0, p)
    else:
        raise ValueError("alternative debe ser 'greater', 'less' o 'two-sided'.")

    return float(p)

# Ejemplo de uso:
x = np.random.normal(0, 1.0, 50)
y = np.random.normal(0, 1.2, 60)
print(ftest_var_pvalue(x, y, alternative="two-sided"))
print(ftest_var_pvalue(x, y, alternative="greater"))  # H1: var(x) > var(y)
print(ftest_var_pvalue(x, y, alternative="less"))    # H1: var(x) < var(y)
````

### 3.3.1. ðŸ“˜ Ejemplo 1: ComparaciÃ³n de concentraciÃ³n de thiol en sangre

Se tienen dos grupos de voluntarios:  
- **Grupo Normal** (n=7)  
- **Grupo con Artritis** (n=6)  

Se midiÃ³ la **concentraciÃ³n de thiol (mM)** en sangre para cada individuo.  
El objetivo es determinar si existen **diferencias significativas en las medias** de ambos grupos.

---

#### ðŸ“Œ Paso 1: HipÃ³tesis

- **HipÃ³tesis nula (Hâ‚€):**  
  Las medias de los dos grupos son iguales.  
  $ \mu_{normal} = \mu_{artritis} $

- **HipÃ³tesis alternativa (Hâ‚):**  
  Las medias de los dos grupos son diferentes.  
  $ \mu_{normal} \neq \mu_{artritis} $

---

#### ðŸ“Œ Paso 2: VerificaciÃ³n de igualdad de varianzas (Prueba F de Fisher)
$s_x=0.0756, s_y=0.4405$
- EstadÃ­stico F calculado:  
  $ F = \dfrac{s_{mayor}^2}{s_{menor}^2} =0.1940/0.0057$

- Con los datos:  
  $ F \approx 33.9 $

- p-value (prueba F): muy pequeÃ±o ($0.000251 < 0.05$)  
  â†’ **Se rechaza Hâ‚€ de igualdad de varianzas**.  
  Por lo tanto, se debe usar la prueba t de Student con **varianzas diferentes** (Welch).

---

#### ðŸ“Œ Paso 3: Prueba t de Student (muestras independientes)

- EstadÃ­stico t:  
  $ t \approx -8.3 $  

- Grados de libertad aproximados (Welch):  
  $ \nu \approx 5 $  

- p-value: $0.000294 < 0.01$  

Con un nivel de confianza del **99% ($\alpha = 0.01$)**:  
â†’ **Se rechaza Hâ‚€**  
â†’ Existe una diferencia estadÃ­sticamente significativa entre las medias de los dos grupos.

---

#### ðŸ“Œ InterpretaciÃ³n

La concentraciÃ³n media de thiol en sangre es **significativamente diferente en el grupo con artritis**.

Nota: tiol en sangre cumple un papel clave como antioxidante natural y protege las cÃ©lulas del daÃ±o oxidativo


````python
import numpy as np
import scipy.stats as stats

# ---------------------------
# Datos de la tabla
# ---------------------------
normal = np.array([1.84, 1.92, 1.94, 1.92, 1.85, 1.91, 2.07])
artritis = np.array([2.81, 4.06, 3.62, 3.27, 3.27, 3.76])

# TamaÃ±os
n_x = len(normal)
n_y = len(artritis)

# Medias
x_mean = np.mean(normal)
y_mean = np.mean(artritis)

# Desviaciones estÃ¡ndar
s_x = np.std(normal, ddof=1)
s_y = np.std(artritis, ddof=1)

print("Resumen estadÃ­stico")
print(f"n_x={n_x}, n_y={n_y}")
print(f"media normal = {x_mean:.4f}, media artritis = {y_mean:.4f}")
print(f"s_x={s_x:.4f}, s_y={s_y:.4f}")
print()

# ---------------------------
# 1. Prueba F de Fisher (igualdad de varianzas)
# ---------------------------
# F = mayor varianza / menor varianza
if s_x**2 >= s_y**2:
    F = s_x**2 / s_y**2
    v1, v2 = n_x - 1, n_y - 1
else:
    F = s_y**2 / s_x**2
    v1, v2 = n_y - 1, n_x - 1

# p-value (cola superior)
p_value_F = 1 - stats.f.cdf(F, v1, v2)

print("Prueba F de Fisher")
print(f"EstadÃ­stico F = {F:.4f}")
print(f"gl1={v1}, gl2={v2}")
print(f"p-value F = {p_value_F:.6f}")

if p_value_F <= 0.05:
    print("â†’ Se rechaza H0: varianzas diferentes")
    equal_var = False
else:
    print("â†’ No se rechaza H0: se asumen varianzas iguales")
    equal_var = True
print()

# ---------------------------
# 2. Prueba t de Student (independientes)
# ---------------------------
t_stat, p_value_t = stats.ttest_ind(normal, artritis, equal_var=equal_var)

print("Prueba t de Student")
print(f"EstadÃ­stico t = {t_stat:.4f}")
print(f"p-value t = {p_value_t:.6f}")

alpha = 0.01  # nivel de significancia (99%)
if p_value_t <= alpha:
    print(f"â†’ Con Î±={alpha}, se rechaza H0: medias diferentes")
else:
    print(f"â†’ Con Î±={alpha}, no se rechaza H0: medias iguales")
````

##3.4. ANOVA (AnÃ¡lisis de Varianza) y F-Test

El **AnÃ¡lisis de Varianza (ANOVA)** es una extensiÃ³n de la prueba F que permite evaluar si existen **diferencias significativas entre las medias de tres o mÃ¡s grupos** mediante una sola prueba, controlando el nivel de error.

---

### ðŸ“Œ Â¿QuÃ© evalÃºa ANOVA?

ANOVA compara:

- **Variabilidad entre los grupos**: cuÃ¡nto difieren las medias de los grupos entre sÃ­.  
- **Variabilidad dentro de los grupos**: dispersiÃ³n de los datos en cada grupo.  

Si la variaciÃ³n **entre grupos** es significativamente mayor que la variaciÃ³n **dentro de los grupos**, se concluye que **al menos un grupo es diferente**.

---

### ðŸ“Œ HipÃ³tesis

>Probar si las muestras provienen de una misma poblaciÃ³n (sin diferencias reales en las medias).

- **Hâ‚€:** Todas las medias son iguales  
  $ \mu_1 = \mu_2 = \cdots = \mu_k $  
- **Hâ‚:** Al menos una media es diferente

---

### EstadÃ­stico F en ANOVA

El valor de F se obtiene al comparar la variaciÃ³n entre los grupos con la variaciÃ³n dentro de los grupos:

$$
F = \dfrac{MS_{entre}}{MS_{dentro}} =
\dfrac{SS_{entre} / df_{entre}}{SS_{dentro} / df_{dentro}}
$$

Donde:

- $SS_{entre}$: suma de cuadrados entre grupos  
- $SS_{dentro}$: suma de cuadrados dentro de los grupos  
- $df_{entre} = k - 1$: grados de libertad entre grupos  
- $df_{dentro} = N - k$: grados de libertad dentro de los grupos  
- $k$: nÃºmero de grupos  
- $N$: total de observaciones

---

### ðŸ“Œ DecisiÃ³n

- El valor de **F** se compara con la distribuciÃ³n F $(df_{entre}, df_{dentro})$.  
- Si el valor-p â‰¤ Î±: se **rechaza Hâ‚€** â†’ eso significa que al menos un grupo es distinto.  
- Si el valor-p > Î±: **no se rechaza Hâ‚€**, eso significa que se acepta H0, que dice no hay evidencia de diferencias.

---

###3.4.1 ðŸ“˜ **Ejemplo 1 de ANOVA**
Simulamos calificaciones de alumnos usando tres mÃ©todos de estudio distintos: tradicional, interactivo y con IA.

````python
import numpy as np
from scipy.stats import f_oneway
import matplotlib.pyplot as plt

# Simulamos calificaciones de alumnos usando tres mÃ©todos distintos
np.random.seed(42)
grupo_A = np.random.normal(loc=75, scale=5, size=30)  # MÃ©todo tradicional
grupo_B = np.random.normal(loc=78, scale=5, size=30)  # MÃ©todo interactivo
grupo_C = np.random.normal(loc=82, scale=5, size=30)  # MÃ©todo con IA

# EstadÃ­stica ANOVA
f_stat, p_value = f_oneway(grupo_A, grupo_B, grupo_C)

print(f"EstadÃ­stico F: {f_stat:.4f}")
print(f"Valor-p: {p_value:.4f}")

if p_value < 0.05:
    print("Se rechaza Hâ‚€: hay diferencias significativas entre los grupos.")
else:
    print("No se rechaza Hâ‚€: no hay evidencia de diferencia entre los grupos.")

# VisualizaciÃ³n rÃ¡pida
plt.boxplot([grupo_A, grupo_B, grupo_C], labels=["Tradicional", "Interactivo", "IA"])
plt.title("ComparaciÃ³n de mÃ©todos de enseÃ±anza (ANOVA)")
plt.ylabel("Calificaciones")
plt.grid(True)
plt.show()
````

####3.4.1.1.  AnÃ¡lisis post hoc: Tukey HSD

Cuando ANOVA detecta diferencias significativas entre grupos, es necesario realizar comparaciones mÃºltiples para identificar **cuÃ¡les grupos difieren entre sÃ­**.

El anÃ¡lisis **Tukey HSD** (Honestly Significant Difference) permite comparar **todas las posibles parejas de grupos** controlando el error tipo I.

**Â¿CuÃ¡ndo usar Tukey HSD?**

- Solo si el resultado de ANOVA es significativo (p < Î±)
- Los grupos deben ser independientes y tener varianzas similares

**Este anÃ¡lisis reporta:**
- Las diferencias entre cada par de medias
- El intervalo de confianza para cada diferencia
- Si la diferencia es estadÃ­sticamente significativa


````python
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Combinamos los datos en un DataFrame
datos = np.concatenate([grupo_A, grupo_B, grupo_C])
metodos = (["A"] * len(grupo_A)) + (["B"] * len(grupo_B)) + (["C"] * len(grupo_C))
df = pd.DataFrame({"Calificacion": datos, "Metodo": metodos})

# Aplicar Tukey HSD
tukey = pairwise_tukeyhsd(endog=df["Calificacion"],
                          groups=df["Metodo"],
                          alpha=0.05)

print(tukey)
tukey.plot_simultaneous()
plt.title("Tukey HSD - ComparaciÃ³n entre mÃ©todos")
plt.show()
````

###3.4.2  ðŸ“˜ Ejemplo ANOVA de una vÃ­a â€“ Pureza de cloruro de sodio

Determinar si existen **diferencias significativas en la pureza (%)** del cloruro de sodio en **cinco muestras (A, B, C, D, E)** tomadas de distintas partes de un barril.

---

### ðŸ“Š HipÃ³tesis

- **Hâ‚€:** Las muestras provienen de una misma poblaciÃ³n (todas las medias son iguales).  
  $ \mu_A = \mu_B = \mu_C = \mu_D = \mu_E $

- **Hâ‚:** Al menos una muestra es diferente.

---

### ðŸ“ EstadÃ­stico F (calculado por ANOVA)

Python calcula la razÃ³n de varianzas:

$$
F = \frac{MS_{entre}}{MS_{dentro}} = \frac{1.96}{0.0653} \approx 30.0
$$

- Grados de libertad entre grupos: $df_1 = k - 1 = 5 - 1 = 4$  
- Grados de libertad dentro de grupos: $df_2 = N - k = 20 - 5 = 15$


### ðŸ“ˆ Resultado en Python



````python
import numpy as np
from scipy import stats

# ---------------------------
# Datos: Pureza (%) de cloruro de sodio en 5 muestras del barril
# ---------------------------
A = np.array([98.8, 98.7, 98.9, 98.8])
B = np.array([99.3, 98.7, 98.8, 99.2])
C = np.array([98.3, 98.5, 98.8, 98.8])
D = np.array([98.0, 97.7, 97.4, 97.3])
E = np.array([99.3, 99.4, 99.9, 99.4])

# ---------------------------
# ANOVA de una vÃ­a (One-way ANOVA)
# ---------------------------
F_stat, p_value = stats.f_oneway(A, B, C, D, E)

print("ANOVA de una vÃ­a (pureza de NaCl en diferentes muestras del barril):")
print(f"EstadÃ­stico F = {F_stat:.3f}")
print(f"Valor-p = {p_value:.6f}")

# Nivel de significancia
alpha = 0.05
if p_value <= alpha:
    print(f"â†’ Rechazamos Hâ‚€ (Î±={alpha}): existen diferencias significativas entre las muestras.")
else:
    print(f"â†’ No se rechaza Hâ‚€ (Î±={alpha}): no hay diferencias significativas entre las muestras.")
````

#### âœ… ConclusiÃ³n del ejemplo anterior

Dado que **p < 0.05**, se **rechaza Hâ‚€**.

ðŸ‘‰ Existe una diferencia significativa entre al menos una de las cinco muestras,  
lo que indica que **la pureza del cloruro de sodio no es uniforme en todo el barril.**

##3.5.  ANOVA de dos vÃ­as (Two-Way ANOVA)

La ANOVA de dos vÃ­as se utiliza para analizar **el efecto de dos factores categÃ³ricos** (independientes) sobre una variable cuantitativa dependiente. AdemÃ¡s, permite examinar si existe una **interacciÃ³n** entre esos factores.

**Â¿CuÃ¡ndo se usa?**

Cuando deseamos saber:
1. Si **cada factor individualmente** tiene un efecto significativo sobre la variable de interÃ©s
2. Si **la combinaciÃ³n de ambos factores** (interacciÃ³n) tambiÃ©n tiene un efecto



###3.5.1 ðŸ“˜ Ejemplo ANOVA de dos  vÃ­as

Evaluar si el rendimiento de estudiantes depende de:
- El mÃ©todo de estudio (tradicional, visual, prÃ¡ctico)  
- El turno (maÃ±ana o tarde)

Y tambiÃ©n analizar si el **efecto del mÃ©todo cambia segÃºn el turno**.

**HipÃ³tesis evaluadas**

Para cada factor (A, B) y la interacciÃ³n (AÃ—B):
- Hâ‚€: No hay efecto del factor (o de la interacciÃ³n)
- Hâ‚: SÃ­ hay efecto significativo

Se realizan pruebas F separadas:
- Una por cada factor
- Una para la interacciÃ³n


````python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Simulamos datos
np.random.seed(42)
n = 10  # muestras por grupo

# Factores
metodo = ["Tradicional", "Visual", "PrÃ¡ctico"]
turno = ["MaÃ±ana", "Tarde"]

# Crear combinaciones
data = []
for m in metodo:
    for t in turno:
        calificaciones = np.random.normal(loc=75 + 3 * (m == "PrÃ¡ctico") - 2 * (t == "Tarde"), scale=5, size=n)
        for val in calificaciones:
            data.append([val, m, t])

df = pd.DataFrame(data, columns=["Calificacion", "Metodo", "Turno"])

# Modelo ANOVA de dos vÃ­as
modelo = ols("Calificacion ~ C(Metodo) + C(Turno) + C(Metodo):C(Turno)", data=df).fit()
anova_table = sm.stats.anova_lm(modelo, typ=2)

print(anova_table)
````

>El turno tambiÃ©n influye significativamente. Las calificaciones promedio son distintas entre maÃ±ana y tarde.

3. InteracciÃ³n (Metodo Ã— Turno) (p = 0.0093 < 0.05):
>Existe una interacciÃ³n significativa.
Esto significa que el efecto del mÃ©todo depende del turno.
Ejemplo: un mÃ©todo puede funcionar mejor en la maÃ±ana que en la tarde.

* Efecto principal del mÃ©todo: Â¿hay diferencias entre mÃ©todos, sin importar el turno?

* Efecto principal del turno: Â¿hay diferencia entre maÃ±ana y tarde, en general?

###3.5.2. ðŸ“˜ Â¿Por quÃ© se usa OLS antes de ANOVA?
En la prÃ¡ctica, se implementa ajustando un modelo de **mÃ­nimos cuadrados ordinarios (OLS, Ordinary Least Squares)**, y luego se descompone la variabilidad en **fuentes de variaciÃ³n** (factores, interacciÃ³n y error).

---

**ðŸ”¹ Â¿QuÃ© hace OLS en este contexto?**

1. **Representa el diseÃ±o experimental como un modelo lineal:**

  $
   Y_{ijk} = \mu + \alpha_i + \beta_j + (\alpha\beta)_{ij} + \varepsilon_{ijk}
  $

   - $\mu$: media general  
   -  $\alpha_i$: efecto del factor A (ej. MÃ©todo)  
   - $\beta_j$: efecto del factor B (ej. Turno)  
   - $(\alpha\beta)_{ij}$: interacciÃ³n entre factores  
   - $\varepsilon_{ijk}$: error aleatorio  

2. **Estima las medias de cada grupo** usando regresiÃ³n lineal.  
   En un diseÃ±o balanceado, los **ajustes de OLS coinciden con las medias crudas** de cada grupo.

3. **Calcula sumas de cuadrados (SS):**  
   - VariaciÃ³n **entre grupos** (debida a los factores e interacciÃ³n).  
   - VariaciÃ³n **dentro de grupos** (error).  

4. **Construye los estadÃ­sticos F** a partir de esas sumas de cuadrados, que son la base del ANOVA.

---

### ðŸ“˜ En resumen

- OLS **traduce el diseÃ±o en un modelo matemÃ¡tico**.  
- ANOVA usa ese modelo para **particionar la variabilidad total** y decidir si los factores tienen un efecto significativo.  
- Sin OLS, solo tendrÃ­amos medias; con OLS, tenemos ademÃ¡s la **variabilidad dentro de los grupos**, necesaria para el contraste estadÃ­stico.

##4.1. ðŸ” Prueba de Dixon (Q-test)

Se usa en **muestras pequeÃ±as** para detectar si el valor mÃ¡s extremo (mÃ¡ximo o mÃ­nimo) es un outlier.

$
Q = \frac{|x_{\text{sospechoso}} - x_{\text{vecino}}|}{x_{\text{max}} - x_{\text{min}}}
$

- Si $Q_{calc} > Q_{crit} $, el valor extremo se considera **atÃ­pico**.
- Los valores crÃ­ticos $ Q_{crit} $ dependen del tamaÃ±o de la muestra (n) y del nivel de confianza (Î±).

âš ï¸ El Q-test solo detecta **un outlier a la vez** y supone distribuciÃ³n **aproximadamente normal**.

---

**Referencias:**
* [Critical Values for Six Dixon Tests](https://www.researchgate.net/publication/28113620_Critical_values_for_six_Dixon_tests_for_outliers_in_normal_samples_up_to_sizes_100_and_applications_in_science_and_engineering)
* [Dixon Test N7 (Q)](http://www.statistics4u.com/fundstat_eng/cc_outlier_tests_dixon.html)

### 4.1.1. ðŸ“˜Ejemplo 1 de Dixon

````python
# --- Ejemplo: Prueba de Dixon Q para detectar outliers en una muestra pequeÃ±a ---

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# FunciÃ³n Q-test para el valor mÃ¡s pequeÃ±o
def q_test_for_smallest_point(dataset):
    dataset = np.sort(dataset)
    return (dataset[1] - dataset[0]) / (dataset[-1] - dataset[0])

# Simular datos con una distribuciÃ³n normal (sin outliers)
np.random.seed(1984)
data_norm = norm.rvs(size=50)

# Agregar un valor extremo artificial
data_norm = np.append(data_norm, -2.5)

# Calcular el estadÃ­stico Q
q_calc = q_test_for_smallest_point(data_norm)
alpha = 0.05
Q_crit = 0.2216  # valor crÃ­tico (para n=50, Î±=0.05)

print(f"EstadÃ­stico Q calculado = {q_calc:.3f}")
print(f"Valor crÃ­tico (Î±={alpha}) = {Q_crit}")
if q_calc > Q_crit:
    print("â†’ El valor mÃ¡s pequeÃ±o es un posible outlier (rechazar Hâ‚€).")
else:
    print("â†’ No hay evidencia suficiente para considerarlo outlier.")

# --- VisualizaciÃ³n ---
plt.figure(figsize=(7,4))
ax = sns.histplot(data_norm, kde=True, color="skyblue", bins=15)
plt.title("DistribuciÃ³n de datos con posible outlier (Q-test)")
plt.axvline(x=np.min(data_norm), color="red", linestyle="--", label="Valor mÃ¡s pequeÃ±o")
plt.legend()
plt.show()
````

##4.2. Prueba de Dixon (Q-test) â€“ detecciÃ³n de valores extremos

El **test de Dixon** puede aplicarse para detectar **outliers en ambos extremos** del conjunto de datos.

- **Valor sospechoso inferior (mÃ­nimo):**
 $  Q_{\text{min}} = \frac{x_2 - x_1}{x_{\text{max}} - x_{\text{min}}}  $
- **Valor sospechoso superior (mÃ¡ximo):**
$  Q_{\text{max}} = \frac{x_{\text{max}} - x_{n-1}}{x_{\text{max}} - x_{\text{min}}} $

Donde $ x_1 $ y $ x_n $ son los valores mÃ¡s extremos.

> Si $Q_{calc} > Q_{crit} $, el valor se considera **atÃ­pico (outlier)**.

**Notas:**
- Adecuado para muestras pequeÃ±as (n â‰¤ 30).  
- Asume distribuciÃ³n aproximadamente normal.  
- Detecta **un solo outlier a la vez**.  

ðŸ“Ž **Referencias:**
* [Critical Values for Six Dixon Tests](https://www.researchgate.net/publication/28113620_Critical_values_for_six_Dixon_tests_for_outliers_in_normal_samples_up_to_sizes_100_and_applications_in_science_and_engineering)
* [Dixon Test N7 (Q)](http://www.statistics4u.com/fundstat_eng/cc_outlier_tests_dixon.html)


### 4.1.2. ðŸ“˜ Ejemplo 2 de Dixon

````python
# --- Prueba de Dixon (Q-test) para detectar outliers en ambos extremos ---

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Funciones Q-test
def q_test_for_smallest_point(dataset):
    dataset = np.sort(dataset)
    return (dataset[1] - dataset[0]) / (dataset[-1] - dataset[0])

def q_test_for_largest_point(dataset):
    dataset = np.sort(dataset)
    return (dataset[-1] - dataset[-2]) / (dataset[-1] - dataset[0])

# --- Simular datos ---
np.random.seed(1984)
data_norm = norm.rvs(size=20)

# Agregar outliers artificiales
#data_norm = np.append(data_norm, [5.5])
#data_norm = np.append(data_norm, [-4.8])
data_norm = np.append(data_norm, [-4.8, 5.5])

# --- ParÃ¡metros del test ---
alpha = 0.05
Q_crit = 0.412  # valor crÃ­tico para n=22 (aprox) y Î±=0.05

# --- Calcular los estadÃ­sticos Q ---
q_min = q_test_for_smallest_point(data_norm)
q_max = q_test_for_largest_point(data_norm)

print(f"EstadÃ­stico Q (menor) = {q_min:.3f}")
print(f"EstadÃ­stico Q (mayor) = {q_max:.3f}")
print(f"Valor crÃ­tico (Î±={alpha}) = {Q_crit}\n")

# --- InterpretaciÃ³n ---
if q_min > Q_crit:
    print("â†’ El valor mÃ­nimo es un posible outlier (rechazar Hâ‚€).")
else:
    print("â†’ El valor mÃ­nimo no es significativamente atÃ­pico.")

if q_max > Q_crit:
    print("â†’ El valor mÃ¡ximo es un posible outlier (rechazar Hâ‚€).")
else:
    print("â†’ El valor mÃ¡ximo no es significativamente atÃ­pico.")

# --- VisualizaciÃ³n ---
plt.figure(figsize=(8,4))
sns.histplot(data_norm, kde=True, bins=15, color="skyblue")
plt.title("DetecciÃ³n de outliers con la prueba de Dixon (Q-test)")
plt.axvline(np.min(data_norm), color="red", linestyle="--", label="Valor mÃ­nimo")
plt.axvline(np.max(data_norm), color="purple", linestyle="--", label="Valor mÃ¡ximo")
plt.legend()
plt.show()
````

##6.1. Prueba Chi-cuadrada (Ï‡Â²) â€” Aplicaciones en pruebas de hipÃ³tesis

La **prueba Chi-cuadrada (Ï‡Â²)** se utiliza para comparar frecuencias **observadas** y **esperadas** en datos categÃ³ricos.  
Permite evaluar si las diferencias observadas se deben al azar o si reflejan un efecto real.

La distribuciÃ³n Chi-cuadrada se usa aquÃ­ como **distribuciÃ³n de referencia** para calcular probabilidades y valores crÃ­ticos.

---

**ðŸ”¹ Tipos de pruebas Chi-cuadrada**

Existen tres aplicaciones principales de la prueba Ï‡Â², dependiendo del objetivo del anÃ¡lisis.

---

**1ï¸âƒ£ Prueba de bondad de ajuste**

EvalÃºa quÃ© tan bien una muestra de datos categÃ³ricos **se ajusta a una distribuciÃ³n teÃ³rica**.

**HipÃ³tesis:**
- $H_0$: La distribuciÃ³n observada se ajusta al modelo teÃ³rico.
- $H_1$: La distribuciÃ³n observada no se ajusta al modelo teÃ³rico.

**EstadÃ­stico:**
$ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} $

donde:
- $O_i$: frecuencia observada en la categorÃ­a *i*  
- $E_i$: frecuencia esperada bajo $H_0$

> **Ejemplo:**  
Comprobar si un **dado es justo** (todas las caras tienen igual probabilidad).

---

**2ï¸âƒ£ Prueba de independencia**

Determina si dos variables categÃ³ricas estÃ¡n relacionadas en una **tabla de contingencia**.

**HipÃ³tesis:**
- $H_0$: Las variables son independientes.  
- $H_1$: Las variables no son independientes.

**EstadÃ­stico:**
$ \chi^2 = \sum_{i=1}^{r} \sum_{j=1}^{c} \frac{(O_{ij} - E_{ij})^2}{E_{ij}} $

donde:
- $O_{ij}$: frecuencia observada en la celda (i, j)  
- $E_{ij} = \dfrac{(F_{i\cdot})(F_{\cdot j})}{N}$: frecuencia esperada  
- $r$, $c$: nÃºmero de filas y columnas  
- $N$: total de observaciones  

> **Ejemplo:**  
Analizar si el **voto** (candidato) depende del **gÃ©nero** del votante.

---

**3ï¸âƒ£ Prueba de homogeneidad**

Verifica si **varios grupos** tienen la **misma distribuciÃ³n de categorÃ­as**.

**HipÃ³tesis:**
- $H_0$: Las proporciones son iguales entre los grupos.  
- $H_1$: Al menos un grupo difiere en sus proporciones.

**EstadÃ­stico:**
$ \chi^2 = \sum_{i=1}^{r} \sum_{j=1}^{c} \frac{(O_{ij} - E_{ij})^2}{E_{ij}} $

âž¡ï¸ La fÃ³rmula es la misma que en la prueba de independencia,  
pero aquÃ­ las **filas representan diferentes poblaciones** o tratamientos.

> **Ejemplo:**  
Comparar si tres marcas de detergente tienen igual aceptaciÃ³n entre consumidores.

---

**ðŸ”¹ InterpretaciÃ³n general**

- Si el valor calculado $ \chi^2_{calc} $ es **mayor** que el valor crÃ­tico $ \chi^2_{crit} $,  
  o si el **p-value < Î±**, entonces **se rechaza Hâ‚€**.  
- En caso contrario, **no hay evidencia suficiente** para rechazar Hâ‚€.

---

**ðŸ“˜ Resumen de modalidades Chi-cuadrada**

| Tipo de prueba     | PropÃ³sito principal                              | Datos requeridos             | EstadÃ­stico base                      |
|--------------------|--------------------------------------------------|------------------------------|--------------------------------------|
| Bondad de ajuste   | Ajuste de datos a una distribuciÃ³n teÃ³rica       | Frecuencias por categorÃ­a     | $ \chi^2 = \sum \frac{(O - E)^2}{E} $ |
| Independencia      | RelaciÃ³n entre dos variables categÃ³ricas         | Tabla de contingencia         | $ \chi^2 = \sum_{i,j} \frac{(O_{ij} - E_{ij})^2}{E_{ij}} $ |
| Homogeneidad       | Igualdad de distribuciones entre grupos          | Varias muestras categÃ³ricas   | $ \chi^2 = \sum_{i,j} \frac{(O_{ij} - E_{ij})^2}{E_{ij}} $ |



###6.1.1. ðŸŽ² Ejemplo - Ï‡Â² de bondad de ajuste
En este ejemplo, simulamos lanzamientos de dados para verificar si los resultados observados
siguen una **distribuciÃ³n uniforme**, como se esperarÃ­a para un dado justo.

**HipÃ³tesis:**
- $H_0$: las frecuencias observadas se ajustan a una distribuciÃ³n uniforme (un dado justo)
- $H_1$: las frecuencias difieren de la distribuciÃ³n uniforme (dado sesgado)

**EstadÃ­stico:**
$ \chi^2 = \sum \frac{(O - E)^2}{E} $

donde $O$ son las frecuencias observadas y $E$ las esperadas.

**InterpretaciÃ³n:**
- Si el valor $p < Î±$ â†’ se **rechaza Hâ‚€** (el dado podrÃ­a estar sesgado)
- Si $p > Î±$ â†’ **no hay evidencia suficiente** para rechazar Hâ‚€ (dado justo)

El ejemplo tambiÃ©n muestra cÃ³mo al **aumentar el tamaÃ±o de muestra** mejora la aproximaciÃ³n
a la distribuciÃ³n teÃ³rica uniforme, reduciendo la variabilidad y aumentando la precisiÃ³n del test.


#### Prueba Chi-cuadrada de bondad de ajuste: Â¿el dado es justo?

Queremos saber si un **dado es justo**, es decir, si cada cara (1, 2, 3, 4, 5, 6) aparece con la **misma probabilidad**.

### HipÃ³tesis

- **Hâ‚€ (nula):** el dado es justo  
  $ P(1)=P(2)=\cdots=P(6)=\frac{1}{6}$

- **Hâ‚ (alternativa):** el dado **no** es justo  
  (al menos una cara tiene probabilidad distinta).

### Idea de la prueba

1. Lanzamos el dado \(n\) veces y contamos cuÃ¡ntas veces aparece cada cara â†’ **frecuencias observadas** $O_i$.
2. Si el dado fuera justo, cada cara deberÃ­a aparecer aproximadamente $E_i = n/6$ veces â†’ **frecuencias esperadas**.
3. Calculamos el estadÃ­stico Chi-cuadrada:
   $   \chi^2 = \sum_{i=1}^{6} \frac{(O_i - E_i)^2}{E_i}$
4. Con ese $\chi^2$ y los grados de libertad $(gl = 6 - 1 = 5$), obtenemos el **valor p**.
---
 Criterio de decisiÃ³n (nivel de significancia $\alpha = 0.05$)

- Si el **valor p < 0.05** â†’ rechazamos Hâ‚€ â†’ los datos **no** son compatibles con un dado justo.
- Si el **valor p â‰¥ 0.05** â†’ no rechazamos Hâ‚€ â†’ los datos son **compatibles** con un dado justo.


````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --- Lanzamientos de un dado ---
np.random.seed(0)
n = 120
resultados = np.random.randint(1, 7, size=n)

# Frecuencias observadas
obs = np.bincount(resultados, minlength=7)[1:]   # omitir Ã­ndice 0
esp = np.full(6, n/6)  # dado justo

print("Frecuencias observadas:", obs)
print("Frecuencias esperadas:", esp)

# Prueba Chi-cuadrada
chi2, p = stats.chisquare(obs, esp)
print(f"\nEstadÃ­stico Ï‡Â²: {chi2:.4f}")
print(f"Valor p: {p:.4f}")

alpha = 0.05
if p < alpha:
    print("\nConclusiÃ³n: El dado NO parece ser justo.")
else:
    print("\nConclusiÃ³n: Los datos son compatibles con un dado justo.")

# GrÃ¡fica
caras = np.arange(1, 7)
plt.bar(caras - 0.15, obs, width=0.3, label="Observada")
plt.bar(caras + 0.15, esp, width=0.3, label="Esperada (justo)")
plt.xticks(caras)
plt.xlabel("Cara del dado")
plt.ylabel("Frecuencia")
plt.title("Bondad de ajuste: Dado justo")
plt.legend()
plt.show()
````

### 6.1.4. Prueba Exacta de Fisher (alternativa para prueba de independencia)

La **prueba exacta de Fisher** evalÃºa si existe una **asociaciÃ³n** entre dos variables categÃ³ricas en una **tabla de contingencia 2Ã—2**.  
Es una alternativa mÃ¡s precisa que la prueba de chi-cuadrado cuando:

- Las frecuencias esperadas son bajas (< 5)
- El tamaÃ±o muestral es pequeÃ±o (< 1000)
- Se requiere una prueba exacta sin aproximaciones

---
HipÃ³tesis
- **Hâ‚€:** Las dos variables son independientes.  
- **Hâ‚:** Las variables estÃ¡n asociadas.

---
Fisher mostrÃ³ que la probabilidad de obtener los datos observados se calcula usando la **distribuciÃ³n hipergeomÃ©trica**:

$
p = \frac{\binom{a+b}{a}\binom{c+d}{c}}{\binom{n}{a+c}}
$

donde la tabla tiene:

|                | Ã‰xito (sÃ­) | Fracaso (no) |
|----------------|:----------:|:----------:|
| **Grupo 1**    | a          | b          |
| **Grupo 2**    | c          | d          |

 $ n = a+b+c+d$.

La prueba suma la probabilidad de la tabla observada mÃ¡s todas las **tablas con desviaciÃ³n igual o mayor** en la misma direcciÃ³n.

---

## ðŸ¿ï¸ Ejemplo: Ardillas y el sonido de alarma

En un estudio ecolÃ³gico, se liberaron ardillas a dos distancias de su madriguera (10 m y 100 m):

| Distancia     | Hizo sonido | No hizo sonido |
|---------------|:-----------:|:--------------:|
| **10 m**      | 16          | 8              |
| **100 m**     | 3           | 18             |

Queremos saber si la probabilidad de hacer el sonido de alarma depende de la distancia a la madriguera.

A continuaciÃ³n visualizamos los porcentajes.


````python
import numpy as np
import matplotlib.pyplot as plt

labels = ['A 10 metros', 'A 100 metros']
hizo = [16, 3]
no_hizo = [8, 18]

total_10 = sum([16, 8])
total_100 = sum([3, 18])

hizo_pct = [16 / total_10 * 100, 3 / total_100 * 100]
no_pct   = [8 / total_10 * 100, 18 / total_100 * 100]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(9, 5))
plt.bar(x - width/2, hizo_pct, width, label="Hizo sonido", color="skyblue")
plt.bar(x + width/2, no_pct, width, label="No hizo sonido", color="salmon")

plt.ylabel("Porcentaje (%)")
plt.title("Comportamiento de alarma segÃºn distancia a la madriguera")
plt.xticks(x, labels)
plt.legend()
plt.grid(axis="y", alpha=0.3)
plt.show()
````

Ahora aplicamos la prueba de Fisher para esta tabla:

$
\begin{pmatrix}
16 & 8 \\
3  & 18
\end{pmatrix}
$

La funciÃ³n `fisher_exact()` devuelve la *odds ratio* y el p-valor.


````python
import numpy as np
from scipy.stats import fisher_exact

tabla = np.array([[16, 8],
                  [ 3, 18]])

oddsratio, p_fisher = fisher_exact(tabla, alternative="two-sided")

print("=== Prueba Exacta de Fisher ===")
print(f"Odds ratio = {oddsratio:.3f}")
print(f"p-valor    = {p_fisher:.5f}")
````

#### ðŸ”¢ Odds Ratio (OR)

La **odds ratio** compara la "posibilidad" de que ocurra un evento entre dos grupos.

$\text{OR} = \frac{a/b}{c/d} = \frac{a\cdot d}{b\cdot c}$

InterpretaciÃ³n:

- **OR = 1:** no hay diferencia entre grupos  
- **OR > 1:** el evento es mÃ¡s frecuente en el grupo 1  
- **OR < 1:** el evento es mÃ¡s frecuente en el grupo 2  

En nuestro caso, OR â‰ˆ 12, lo cual indica que las ardillas **a 10 m tienen 12 veces mÃ¡s odds de hacer el sonido** que las ardillas a 100 m.

TambiÃ©n podemos obtener un **intervalo de confianza al 95 %**.

````python
import math

a, b, c, d = tabla.flatten()

log_or = np.log(oddsratio)
se_log_or = math.sqrt(1/a + 1/b + 1/c + 1/d)
z = 1.96

ci_low = np.exp(log_or - z * se_log_or)
ci_high = np.exp(log_or + z * se_log_or)

print(f"IC 95% de la OR: ({ci_low:.2f} , {ci_high:.2f})")
````

#### ComparaciÃ³n con chi-cuadrado de independencia

Aunque la prueba exacta de Fisher es mÃ¡s precisa para muestras pequeÃ±as,  es Ãºtil comparar con chi-cuadrado para ver si las conclusiones coinciden.


````python
from scipy.stats import chi2_contingency
import numpy as np

# Tabla de contingencia (ejemplo ardillas)
tabla = np.array([[16, 8],
                  [ 3, 18]])

# --- Sin correcciÃ³n de Yates ---
chi2_nc, p_nc, dof_nc, expected_nc = chi2_contingency(tabla, correction=False)

# --- Con correcciÃ³n de Yates (solo 2x2) ---
chi2_y, p_y, dof_y, expected_y = chi2_contingency(tabla, correction=True)

print("=== ChiÂ² de independencia ===\n")

print("--- Sin correcciÃ³n de Yates ---")
print(f"ChiÂ² = {chi2_nc:.3f}")
print(f"gl   = {dof_nc}")
print(f"p    = {p_nc:.5f}")
print("Frecuencias esperadas (sin correcciÃ³n):")
print(expected_nc)

print("\n--- Con correcciÃ³n de Yates ---")
print(f"ChiÂ² (Yates) = {chi2_y:.3f}")
print(f"gl           = {dof_y}")
print(f"p (Yates)    = {p_y:.5f}")
print("Frecuencias esperadas (las mismas):")
print(expected_y)

print("\nComparaciÃ³n rÃ¡pida de p-valores:")
print(f"p sin correcciÃ³n = {p_nc:.5f}")
print(f"p con Yates      = {p_y:.5f}")
````

#### CorrecciÃ³n de Yates en la prueba Ï‡Â²

En tablas **2Ã—2**, la prueba de Ï‡Â² usa una **distribuciÃ³n continua** (Ï‡Â²) para aproximar datos de **conteos discretos**.  
Cuando las frecuencias esperadas son **pequeÃ±as**, esa aproximaciÃ³n puede **exagerar** la evidencia contra $H_0$ (p-valores demasiado pequeÃ±os).

La **correcciÃ³n de continuidad de Yates** ajusta el cÃ¡lculo de Ï‡Â² restando 0.5 a la diferencia \(|O - E|\):

$\chi^2_{\text{Yates}} = \sum \frac{(|O - E| - 0.5)^2}{E}$

Esto hace la prueba mÃ¡s **conservadora** (Ï‡Â² menor â†’ p-valor mayor).

**Â¿CuÃ¡ndo aplicarla?**

- Tablas **2Ã—2**  
- Frecuencias esperadas **bajas** (por ejemplo, alguna celda < 5)  
- Cuando se desea ser mÃ¡s **precavido** para no rechazar $H_0$ demasiado fÃ¡cil  

En `scipy.stats.chi2_contingency`:

- `correction=True` â†’ **con** Yates (recomendable en 2Ã—2 con n pequeÃ±o).  
- `correction=False` â†’ versiÃ³n clÃ¡sica **sin** correcciÃ³n.


##6.2. Mannâ€“Whitney U Test

El **Mannâ€“Whitney U test** es una alternativa no paramÃ©trica a la prueba **t de Student para muestras independientes**.  
Se utiliza cuando:
- Los datos no siguen una distribuciÃ³n normal.  
- Las variables son ordinales o los tamaÃ±os de muestra son pequeÃ±os.  

ðŸ”¹**HipÃ³tesis:**
- Hâ‚€: ambas muestras provienen de la misma poblaciÃ³n (no hay diferencia).  
- Hâ‚: las muestras provienen de poblaciones diferentes.
---
ðŸ”¹ **EstadÃ­stico U:**  
El valor **U** cuantifica el **grado de solapamiento entre las distribuciones** de ambos grupos.  
Se obtiene al ordenar todos los valores y comparar sus rangos.

- Si las distribuciones son idÃ©nticas, **U â‰ˆ nâ‚Â·nâ‚‚ / 2** (valor intermedio).  
- Si las distribuciones difieren mucho, **U** serÃ¡ **pequeÃ±o o grande**, dependiendo de la direcciÃ³n de la diferencia.
---
El **valor p** asociado a U indica la probabilidad de observar una diferencia igual o mayor bajo Hâ‚€.

ðŸ”¹**DecisiÃ³n:**
- Si $p < Î±$: se rechaza Hâ‚€ â†’ existe diferencia significativa entre los grupos.  
- Si $p > Î±$: no se rechaza Hâ‚€ â†’ las distribuciones son similares.

---
ðŸ”¹Ejemplo: Se generan dos grupos con tendencia central ligeramente distintas (50 y 51).  
El test evalÃºa si esa diferencia es estadÃ­sticamente significativa sin suponer normalidad.


````python
# --- Prueba de Mannâ€“Whitney U ---
# Alternativa no paramÃ©trica a la t de Student (muestras independientes)

from numpy.random import seed, randn
from scipy.stats import mannwhitneyu
import seaborn as sns
import matplotlib.pyplot as plt

# --- Reproducibilidad ---
seed(1962)

# --- Generamos dos muestras (ligeramente distintas en la media) ---
data1 = 5 * randn(100) + 50   # muestra 1
data2 = 5 * randn(200) + 51   # muestra 2

# --- Prueba de Mannâ€“Whitney U ---
stat, p = mannwhitneyu(data1, data2, alternative='two-sided')

print(f"EstadÃ­stico U = {stat:.3f}")
print(f"Valor p = {p:.5f}")

# --- InterpretaciÃ³n ---
alpha = 0.05
print("\n--- InterpretaciÃ³n ---")
print("Hâ‚€: ambas muestras provienen de la misma distribuciÃ³n.")
print("Hâ‚: las muestras provienen de distribuciones diferentes.")

if p > alpha:
    print(f"No se rechaza Hâ‚€ (p = {p:.3f} > {alpha}) â†’ Misma distribuciÃ³n.")
else:
    print(f"Se rechaza Hâ‚€ (p = {p:.3f} â‰¤ {alpha}) â†’ Distribuciones diferentes.")

# --- VisualizaciÃ³n ---
plt.figure(figsize=(7,4))
sns.kdeplot(data1, fill=True, alpha=0.4, label="Grupo 1")
sns.kdeplot(data2, fill=True, alpha=0.4, label="Grupo 2")
plt.title("ComparaciÃ³n de dos grupos â€” Mannâ€“Whitney U test")
plt.xlabel("Valores observados")
plt.ylabel("Densidad")
plt.legend()
plt.show()
````

## 6.3. Wilcoxon signed-rank
# **Prueba Wilcoxon Signed-Rank**

**Prueba de rangos con signo de Wilcoxon**

Es una prueba estadÃ­stica no paramÃ©trica utilizada para comparar **dos muestras relacionadas o emparejadas**, o para evaluar si la **mediana de las diferencias entre pares de observaciones** es significativamente distinta de cero.

Se usa como **alternativa no paramÃ©trica** a la prueba **t de Student para muestras relacionadas**, cuando no se cumple el supuesto de normalidad en las diferencias.

### **CÃ³mo funciona**

Supongamos que tenemos dos conjuntos de datos emparejados:  
$X_1, X_2, \dots, X_n$ y $Y_1, Y_2, \dots, Y_n$.

1. Calcular las diferencias para cada par:  
   $
   d_i = X_i - Y_i
   $
2. Eliminar las diferencias nulas (donde $d_i = 0$).
3. Tomar el valor absoluto de las diferencias $|d_i|$ y ordenarlas de menor a mayor.
4. Asignar rangos a los valores absolutos (1 al mÃ¡s pequeÃ±o, 2 al siguiente, etc.).  
   En caso de empates, se asigna el **rango promedio**.
5. Asignar signos a los rangos (recuperar el signo original de cada diferencia) y calcular:
   - $R^+$: suma de rangos con diferencia **positiva**  
   - $R^-$: suma de rangos con diferencia **negativa**
6. Calcular el estadÃ­stico de prueba:  
   $
   W = \min(R^+, R^-)
   $
7. Tomar decisiÃ³n:
   - Para muestras pequeÃ±as $(n \leq 25$, despuÃ©s de eliminar ceros):  
     - Comparar \(W\) con valores crÃ­ticos de la **tabla de Wilcoxon**.  
     - Si $W \leq W_{\text{crÃ­tico}}$, se rechaza $H_0$.
   - Para muestras grandes $(n > 25$):  
     - Usar aproximaciÃ³n normal:
       $
       Z = \frac{W - \frac{n(n+1)}{4}}{\sqrt{\frac{n(n+1)(2n+1)}{24}}}
       $
     - Comparar con valores $Z$ de la distribuciÃ³n normal estÃ¡ndar.

### **HipÃ³tesis**

- $H_0$: La **mediana de las diferencias** es igual a 0 (no hay cambio sistemÃ¡tico entre las mediciones pareadas).  
- $H_1$: La **mediana de las diferencias** es distinta de 0 (existe cambio sistemÃ¡tico).

### **Ejemplo paso a paso**

Se busca comparar los niveles de glucosa en sangre de 10 pacientes **antes y despuÃ©s** de seguir una dieta durante 8 semanas.



````python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos
antes = np.array([155, 132, 178, 160, 143, 190, 175, 121, 169, 185])
despues = np.array([140, 128, 178, 145, 158, 185, 162, 150, 155, 185])

datos = pd.DataFrame({'Antes': antes, 'DespuÃ©s': despues})
display(datos)

# Boxplot comparativo
plt.boxplot([datos['Antes'], datos['DespuÃ©s']], labels=['Antes', 'DespuÃ©s'])
plt.title('DistribuciÃ³n de glucosa antes y despuÃ©s de la dieta')
plt.ylabel('Nivel de glucosa (mg/dL)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

print("La mayorÃ­a de los pacientes muestran una reducciÃ³n moderada,\n"
      "pero hay solapamiento entre los niveles de glucosa antes y despuÃ©s.\n")

# =============================================================================
# PASO 1: Calcular las diferencias
# =============================================================================
datos['di'] = datos['Antes'] - datos['DespuÃ©s']
display(datos)

print(
    "- di > 0: la glucosa bajÃ³ despuÃ©s de la dieta (efecto deseado)\n"
    "- di < 0: la glucosa subiÃ³ (la dieta no ayudÃ³ o empeorÃ³)\n"
    "- di = 0: no hubo cambio\n"
)

# =============================================================================
# PASO 2: Eliminar diferencias nulas
# =============================================================================
datos = datos[datos['di'] != 0].copy()
display(datos)

# =============================================================================
# PASO 3: Ordenar por valor absoluto
# =============================================================================
datos['abs_di'] = np.abs(datos['di'])  # valor absoluto de las diferencias
datos = datos.sort_values('abs_di').reset_index(drop=True)  # ordenar
display(datos)

# =============================================================================
# PASO 4: Asignar rangos (con manejo de empates)
# =============================================================================
rangos = []
conteo_empates = datos['abs_di'].value_counts()

for i, valor in enumerate(datos['abs_di']):
    frecuencia = conteo_empates[valor]  # cuÃ¡ntas veces aparece este valor

    # No hay empate
    if frecuencia == 1:
        rangos.append(i + 1)
    # Hay empate: asignar rango promedio a todos los que comparten este valor
    else:
        indices_empate = datos.index[datos['abs_di'] == valor].tolist()
        rango_promedio = np.mean([idx + 1 for idx in indices_empate])
        rangos.append(rango_promedio)

datos['rango'] = rangos
display(datos)

# =============================================================================
# PASO 5: Sumar R+ y R- a partir de los rangos
# =============================================================================
# (opcional) columna con rango firmado, solo para inspecciÃ³n
datos['rango_con_signo'] = datos['rango'] * np.sign(datos['di'])

R_positivo = datos[datos['di'] > 0]['rango'].sum()
R_negativo = datos[datos['di'] < 0]['rango'].sum()

print('R+ =', R_positivo)
print('R- =', R_negativo)
display(datos)

# =============================================================================
# PASO 6: EstadÃ­stico de prueba W (versiÃ³n manual)
# =============================================================================
W_manual = min(R_positivo, R_negativo)
print('W (manual) =', W_manual)

# =============================================================================
# PASO 7: Prueba con SciPy (Wilcoxon pareado)
# =============================================================================
# Usamos los datos emparejados sin las diferencias nulas
# (datos['Antes'] y datos['DespuÃ©s'] ya estÃ¡n filtrados)
statistic, p_value = stats.wilcoxon(datos['Antes'], datos['DespuÃ©s'])

print('\nResultados SciPy:')
print('EstadÃ­stico (suma de rangos positivos) =', statistic)
print('Valor p =', p_value)

# Tomar decisiÃ³n
alpha = 0.05  # nivel de significancia

if p_value < alpha:
    print('\nDECISIÃ“N: Se rechaza Hâ‚€ â†’ '
          'La dieta produjo un cambio significativo en los niveles de glucosa.')
else:
    print('\nDECISIÃ“N: NO se rechaza Hâ‚€ â†’ '
          'No hay evidencia suficiente de un cambio significativo.')
````

## 6.4. Kruskal-Wallis H Test

La **prueba de Kruskalâ€“Wallis H** es la alternativa no paramÃ©trica al **ANOVA de una vÃ­a**.  
Se utiliza para comparar **tres o mÃ¡s grupos independientes** cuando los datos **no siguen una distribuciÃ³n normal**
o cuando las varianzas no son homogÃ©neas.

---

ðŸ”¹**HipÃ³tesis:**
- **Hâ‚€:** todas las muestras provienen de la misma poblaciÃ³n (no hay diferencias significativas).  
- **Hâ‚:** al menos una muestra proviene de una poblaciÃ³n diferente.
---
El **estadÃ­stico H** se calcula a partir de los **rangos promedio** de cada grupo.

---
$
H = \frac{12}{N(N+1)} \sum_{i=1}^{k} n_i \left( R_i - \bar{R} \right)^2
$

donde:

-  $k$: nÃºmero de grupos  
- $N$ : nÃºmero total de observaciones (todos los grupos juntos)  
- $n_i$: tamaÃ±o del grupo \(i\)  
- $R_i$: promedio de rangos del grupo \(i\)  
- $ \bar{R} = \frac{N+1}{2} $: media global de los rangos  

---

#### ðŸ”¹ Forma equivalente (usando la suma total de rangos)

$H = \frac{12}{N(N+1)} \sum_{i=1}^{k} \frac{T_i^2}{n_i} - 3(N+1)$

donde $T_i = \sum R_{ij}$ es la **suma de rangos** del grupo \(i\).

---

---

ðŸ”¹**DecisiÃ³n:**
- Si $p < Î±$ se **rechaza Hâ‚€**; existe diferencia significativa entre grupos.  
- Si $p > Î±$  no se rechaza Hâ‚€; los grupos no difieren significativamente.
---
 **Referencia:**  
Kruskal, W. H., & Wallis, W. A. (1952).  
*Use of ranks in one-criterion variance analysis.*  
*Journal of the American Statistical Association*, 47(260), 583â€“621.




````python
# --- Prueba de Kruskalâ€“Wallis H ---
# Alternativa no paramÃ©trica al ANOVA de una vÃ­a (muestras independientes)

from numpy.random import seed, randn
from scipy.stats import kruskal
import seaborn as sns
import matplotlib.pyplot as plt

# Reproducibilidad
seed(1962)

# --- Generar tres muestras independientes ---
data1 = 5 * randn(100) + 50
data2 = 5 * randn(100) + 50
data3 = 5 * randn(100) + 52

# --- Prueba Kruskalâ€“Wallis ---
stat, p = kruskal(data1, data2, data3)

print(f"EstadÃ­stico H = {stat:.3f}")
print(f"Valor p = {p:.5f}")

# --- InterpretaciÃ³n ---
alpha = 0.05
print("\n--- InterpretaciÃ³n ---")
print("Hâ‚€: todas las muestras provienen de la misma poblaciÃ³n.")
print("Hâ‚: al menos una muestra proviene de una poblaciÃ³n diferente.")

if p > alpha:
    print(f"No se rechaza Hâ‚€ (p = {p:.3f} > {alpha}) â†’ Misma distribuciÃ³n entre grupos.")
else:
    print(f"Se rechaza Hâ‚€ (p = {p:.3f} â‰¤ {alpha}) â†’ Diferencia significativa entre grupos.")

# --- VisualizaciÃ³n ---
plt.figure(figsize=(7,4))
sns.kdeplot(data1, fill=True, alpha=0.4, label="Grupo 1 (media â‰ˆ 50)")
sns.kdeplot(data2, fill=True, alpha=0.4, label="Grupo 2 (media â‰ˆ 50)")
sns.kdeplot(data3, fill=True, alpha=0.4, label="Grupo 3 (media â‰ˆ 52)")
plt.title("ComparaciÃ³n de tres grupos â€” Kruskalâ€“Wallis H-test")
plt.xlabel("Valores observados")
plt.ylabel("Densidad")
plt.legend()
plt.show()
````

````python
import numpy as np
from scipy.stats import shapiro, kruskal, f_oneway
import matplotlib.pyplot as plt

# ---------------------------
# Datos: Pureza (%) de NaCl en 5 muestras del barril
# ---------------------------
A = np.array([98.8, 98.7, 98.9, 98.8])
B = np.array([99.3, 98.7, 98.8, 99.2])
C = np.array([98.3, 98.5, 98.8, 98.8])
D = np.array([98.0, 97.7, 97.4, 97.3])
E = np.array([99.3, 99.4, 99.9, 99.4])

grupos = {"A": A, "B": B, "C": C, "D": D, "E": E}
labels = list(grupos.keys())
values = [grupos[k] for k in labels]

# =====================================================
# 1ï¸âƒ£ Prueba de normalidad (Shapiroâ€“Wilk)
# =====================================================
print("=== Prueba de Normalidad (Shapiroâ€“Wilk) ===")
normal = True
alpha = 0.05

for nombre, datos in grupos.items():
    stat, p = shapiro(datos)
    print(f"{nombre}: W={stat:.3f}, p={p:.4f}", end=" â†’ ")
    if p < alpha:
        print("âŒ No normal")
        normal = False
    else:
        print("âœ… Normal")

# =====================================================
# 2ï¸âƒ£ ANOVA (solo si todos los grupos son normales)
# =====================================================
if normal:
    print("\n=== ANOVA de una vÃ­a ===")
    F_stat, p_value = f_oneway(*values)
    print(f"F = {F_stat:.3f}, p = {p_value:.5f}")
    if p_value <= alpha:
        print(f"â†’ Rechazamos Hâ‚€: existen diferencias significativas.")
    else:
        print(f"â†’ No se rechaza Hâ‚€: no hay diferencias significativas.")
else:
    # =====================================================
    # 3ï¸âƒ£ Kruskalâ€“Wallis (no paramÃ©trica)
    # =====================================================
    print("\n=== Prueba Kruskalâ€“Wallis ===")
    stat, p_value = kruskal(*values)
    print(f"H = {stat:.3f}, p = {p_value:.6f}")
    if p_value <= alpha:
        print(f"â†’ Rechazamos Hâ‚€ (Î±={alpha}): existen diferencias significativas entre las muestras.")
    else:
        print(f"â†’ No se rechaza Hâ‚€ (Î±={alpha}): no hay diferencias significativas entre las muestras.")

# =====================================================
# 4ï¸âƒ£ VisualizaciÃ³n
# =====================================================
plt.figure(figsize=(8, 4.8))
bp = plt.boxplot(values, labels=labels, patch_artist=True, widths=0.6)

for patch in bp['boxes']:
    patch.set_facecolor('#d0e1ff')

# Puntos individuales con jitter
for i, arr in enumerate(values, start=1):
    x = np.random.normal(loc=i, scale=0.04, size=len(arr))
    plt.plot(x, arr, 'o', alpha=0.7, color='steelblue')

plt.title("Pureza de NaCl por muestra (Aâ€“E)")
plt.ylabel("Pureza (%)")

# Mostrar estadÃ­stico y p-valor en la grÃ¡fica
if normal:
    txt = f"ANOVA\nF = {F_stat:.2f}\np = {p_value:.3f}"
else:
    txt = f"Kruskalâ€“Wallis\nH = {stat:.2f}\np = {p_value:.3f}"

plt.text(
    0.02, 0.95,
    txt,
    transform=plt.gca().transAxes,
    va="top",
    ha="left",
    bbox=dict(facecolor="white", alpha=0.8, edgecolor="none")
)
````

## 6.5. Test de Dunn â€” Comparaciones mÃºltiples post hoc

Cuando la **prueba de Kruskalâ€“Wallis** indica diferencias significativas (p < 0.05),  
no especifica **entre quÃ© grupos** se producen esas diferencias.  
Para determinarlo, se realiza una **comparaciÃ³n mÃºltiple** utilizando el **Test de Dunn**.

---

### ðŸ”¹ HipÃ³tesis

Para cada par de grupos (i, j):

- **Hâ‚€:** las medianas de los grupos *i* y *j* son iguales.  
- **Hâ‚:** las medianas de los grupos *i* y *j* son diferentes.

---

### ðŸ”¹ Ajuste por comparaciones mÃºltiples

Dado que se realizan varias comparaciones, se ajustan los valores p mediante mÃ©todos como:
- **Bonferroni** â†’ mÃ¡s conservador (reduce falsos positivos).  
- **Holm** â†’ menos estricto pero controlado.

---

### ðŸ”¹ InterpretaciÃ³n

En la matriz de resultados:
- Cada celda muestra el **valor p ajustado** de la comparaciÃ³n entre dos grupos.  
- Si **p < 0.05**, se concluye que esos grupos **difieren significativamente**.

---

### ðŸ”¹ Ejemplo interpretativo

| ComparaciÃ³n | p-valor ajustado | ConclusiÃ³n |
|--------------|------------------|-------------|
| A vs B | 0.312 | No significativa |
| A vs C | 0.004 | Diferencia significativa |
| B vs C | 0.021 | Diferencia significativa |

âž¡ï¸ Los grupos **C** y **A/B** difieren significativamente,  
mientras que **A y B** no muestran diferencias.

---

## 6.6.  Friedman Test

La **prueba de Friedman** es la alternativa no paramÃ©trica al **ANOVA de medidas repetidas**.
Se utiliza cuando el mismo grupo de sujetos es evaluado **bajo tres o mÃ¡s condiciones** o tratamientos
y no se cumple el supuesto de normalidad.

---

### ðŸ”¹ HipÃ³tesis

- **Hâ‚€:** todas las condiciones o tratamientos producen **efectos similares** (las medianas son iguales).  
- **Hâ‚:** al menos un tratamiento produce un efecto **diferente**.

---

### ðŸ”¹ EstadÃ­stico

El estadÃ­stico de Friedman se calcula sobre los **rangos de cada fila** (participante):

$ \chi^2_F = \frac{12}{n k (k+1)} \sum_{j=1}^{k} R_j^2 - 3n(k+1) $

donde:  
- $n$: nÃºmero de sujetos  
- $k$: nÃºmero de tratamientos  
- $R_j$: suma de rangos por columna (tratamiento)

El valor de $ \chi^2_F $ se contrasta con una distribuciÃ³n **Chi-cuadrada (Ï‡Â²)** con $k - 1$ grados de libertad.

---

### ðŸ”¹ InterpretaciÃ³n

- Si $p < Î±$ â†’ se **rechaza Hâ‚€** â†’ al menos un tratamiento difiere significativamente.  
- Si $p > Î±$ â†’ no se rechaza Hâ‚€ â†’ los tratamientos tienen efectos similares.

---

### ðŸ”¹ AplicaciÃ³n prÃ¡ctica

Esta prueba es muy Ãºtil cuando:
- Se comparan **mÃ©todos, algoritmos o tratamientos** en los **mismos sujetos o muestras**.  
- Los datos no son normales o son **ordinales** (por ejemplo, niveles de satisfacciÃ³n, desempeÃ±o o precisiÃ³n).
---
ðŸ”¹En el ejemplo:
- Cada participante prueba tres mÃ©todos (A, B, C).  
- Los resultados se analizan con la prueba de Friedman para determinar si el rendimiento varÃ­a entre mÃ©todos.

---


````python
# --- Prueba de Friedman con muestras NO normales (exponenciales) ---
# Alternativa no paramÃ©trica al ANOVA de medidas repetidas (datos dependientes)

import numpy as np
from numpy.random import seed
from scipy.stats import friedmanchisquare, shapiro
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Reproducibilidad
seed(1962)

# --- SimulaciÃ³n de datos (NO normales) ---
n = 15  # nÃºmero de sujetos (medidas repetidas)
# Efecto base por sujeto (compartido en los 3 mÃ©todos para inducir correlaciÃ³n intra-sujeto)
base_sujeto = np.random.exponential(scale=3.0, size=n)  # asimÃ©trico

# Ruido asimÃ©trico por mÃ©todo (exponencial) + desplazamientos de "rendimiento"
metodo_A = 70 + base_sujeto + np.random.exponential(scale=2.0, size=n) + 0.0   # MÃ©todo A
metodo_B = 72 + base_sujeto + np.random.exponential(scale=2.5, size=n) + 1.5   # MÃ©todo B
metodo_C = 75 + base_sujeto + np.random.exponential(scale=3.0, size=n) + 3.0   # MÃ©todo C

# --- ComprobaciÃ³n rÃ¡pida de no-normalidad (Shapiro por mÃ©todo) ---
alpha = 0.05
print("=== Shapiroâ€“Wilk por mÃ©todo ===")
for nombre, arr in [("MÃ©todo A", metodo_A), ("MÃ©todo B", metodo_B), ("MÃ©todo C", metodo_C)]:
    W, p = shapiro(arr)
    estado = "No normal âŒ" if p < alpha else "Normal âœ…"
    print(f"{nombre}: W={W:.3f}, p={p:.4f} â†’ {estado}")

# --- Prueba de Friedman ---
stat, p = friedmanchisquare(metodo_A, metodo_B, metodo_C)

print("\n=== Prueba de Friedman ===")
print(f"EstadÃ­stico Ï‡Â² = {stat:.3f}")
print(f"Valor p        = {p:.5f}")

print("\n--- InterpretaciÃ³n ---")
print("Hâ‚€: no hay diferencias entre los tratamientos (medianas iguales).")
print("Hâ‚: al menos un tratamiento difiere en su efecto.")
if p > alpha:
    print(f"No se rechaza Hâ‚€ (p = {p:.3f} > {alpha}) â†’ No hay diferencias significativas.")
else:
    print(f"Se rechaza Hâ‚€ (p = {p:.3f} â‰¤ {alpha}) â†’ Diferencias significativas entre los mÃ©todos.")

# --- VisualizaciÃ³n ---
df = pd.DataFrame({
    "Sujeto": np.arange(1, n+1),
    "MÃ©todo A": metodo_A,
    "MÃ©todo B": metodo_B,
    "MÃ©todo C": metodo_C
})

# Formato largo para seaborn
df_long = df.melt(id_vars="Sujeto", var_name="MÃ©todo", value_name="Puntaje")

plt.figure(figsize=(7.5, 4.8))
sns.boxplot(data=df_long, x="MÃ©todo", y="Puntaje", whis=1.5)
sns.stripplot(data=df_long, x="MÃ©todo", y="Puntaje", alpha=0.7, jitter=0.1)

# Conectar puntos del mismo sujeto para resaltar que son medidas repetidas
# (trazo suave para no saturar)
for sid, sub in df_long.groupby("Sujeto"):
    xs = sub["MÃ©todo"].map({"MÃ©todo A":0, "MÃ©todo B":1, "MÃ©todo C":2}).values
    ys = sub["Puntaje"].values
    plt.plot(xs, ys, alpha=0.25, linewidth=1)

plt.title("ComparaciÃ³n de tres mÃ©todos â€” Prueba de Friedman (muestras NO normales)")
plt.ylabel("Puntaje de rendimiento")
plt.xlabel("")
plt.text(0.02, 0.95, f"Friedman\nÏ‡Â² = {stat:.2f}\np = {p:.3g}",
         transform=plt.gca().transAxes, va="top", ha="left",
         bbox=dict(facecolor="white", alpha=0.85, edgecolor="none"))
plt.tight_layout()
plt.show()
````

## 6.7 Andersonâ€“Darling Test (una muestra)

La **prueba de Andersonâ€“Darling (AÂ²)** es una prueba estadÃ­stica diseÃ±ada para evaluar si un conjunto de datos sigue una **distribuciÃ³n especÃ­fica**, generalmente la **distribuciÃ³n normal**.  
A diferencia de la prueba de Shapiroâ€“Wilk, Andersonâ€“Darling otorga mayor peso a las **colas de la distribuciÃ³n**, lo que la hace mÃ¡s sensible a valores extremos.

---
HipÃ³tesis
 **Hâ‚€ (hipÃ³tesis nula):** Los datos provienen de la distribuciÃ³n especificada.  
- **Hâ‚ (hipÃ³tesis alternativa):** Los datos **no** provienen de esa distribuciÃ³n.

---
Criterio de decisiÃ³n (en tÃ©rminos de *p-value*)

En muchas implementaciones modernas (ej. Python), la decisiÃ³n se toma usando el **valor p (p-value)** asociado al estadÃ­stico AÂ².

Sea Î± el **nivel de significancia** (por ejemplo, Î± = 0.05):

- Si $ \text{p-value} < \alpha$
  â†’ **Se rechaza Hâ‚€**.  
  Concluimos que los datos **no** siguen la distribuciÃ³n especificada (por ejemplo, no son normales).

- Si $\text{p-value} \geq \alpha $
  â†’ **No se rechaza Hâ‚€**.  
  Los datos son **compatibles** con la distribuciÃ³n teÃ³rica.

> En resumen: **p-value pequeÃ±o muestra evidencia contra la normalidad**.
---
EstadÃ­stico Andersonâ€“Darling

Sea $x_{(1)}, x_{(2)}, \dots, x_{(n)}$ la muestra ordenada de menor a mayor, y $F(x)$ la **funciÃ³n de distribuciÃ³n acumulada teÃ³rica** (por ejemplo, la CDF de la normal).  
El estadÃ­stico de Andersonâ€“Darling se define como:
$A^2 = -n - \frac{1}{n} \sum_{i=1}^{n} (2i - 1)
\left[\ln\big(F(x_{(i)})\big) + \ln\big(1 - F(x_{(n+1-i)})\big)\right]$

En palabras:

- Se ordenan los datos.
- Se evalÃºa la CDF teÃ³rica \(F\) en cada dato ordenado.
- Se combinan las contribuciones de la parte baja y alta de la distribuciÃ³n mediante logaritmos.
- Cuanto mayor sea la diferencia entre la distribuciÃ³n observada y la teÃ³rica, **mayor serÃ¡ $A^2$**.

En algunas implementaciones especÃ­ficas (como para la normalidad), se usa una versiÃ³n corregida:

$A^{2*} = A^2 \left(1 + \frac{0.75}{n} + \frac{2.25}{n^2}\right)$

para ajustar por el tamaÃ±o de muestra.

---
Â¿QuÃ© mide el estadÃ­stico AÂ²?

El estadÃ­stico AÂ² compara la **distribuciÃ³n acumulada empÃ­rica** de los datos con la **distribuciÃ³n acumulada teÃ³rica**.  
Cuanto mayor sea la discrepancia entre ambas curvas, mayor serÃ¡ el valor de $A^2$.

---

##  Uso tÃ­pico

La prueba Andersonâ€“Darling se utiliza para:
- Evaluar **normalidad** antes de aplicar pruebas paramÃ©tricas  
- Confirmar supuestos de modelos estadÃ­sticos  
- Verificar si una muestra sigue una distribuciÃ³n especÃ­fica (no solo normal)

Es especialmente Ãºtil cuando se sospecha que la distribuciÃ³n puede tener **colas pesadas** o valores extremos.

---


````python
import numpy as np
import scipy.stats as stats
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. GeneraciÃ³n de datos ejemplo
# -----------------------------
np.random.seed(42)
datos_normales = np.random.normal(loc=50, scale=5, size=100)
datos_no_normales = np.random.exponential(scale=2, size=100)

# -----------------------------
# 2. Prueba Anderson-Darling
# -----------------------------
resultado_normales = stats.anderson(datos_normales, dist='norm')
resultado_no_normales = stats.anderson(datos_no_normales, dist='norm')


# ---------------------------------------------------------
# 3. p-value aproximado para Anderson-Darling (normalidad)
#    FÃ³rmulas estÃ¡ndar usadas en software estadÃ­stico
# ---------------------------------------------------------
def pvalue_anderson_darling(A2):
    if A2 < 0.2:
        p = 1 - np.exp(-13.436 + 101.14*A2 - 223.73*A2**2)
    elif A2 < 0.34:
        p = 1 - np.exp(-8.318 + 42.796*A2 - 59.938*A2**2)
    elif A2 < 0.6:
        p = np.exp(0.9177 - 4.279*A2 - 1.38*A2**2)
    else:
        p = np.exp(1.2937 - 5.709*A2 + 0.0186*A2**2)
    return min(max(p, 0), 1)  # limitar entre 0 y 1


def tabla_anderson_con_pvalue(resultado):
    A2 = resultado.statistic
    p = pvalue_anderson_darling(A2)
    df = pd.DataFrame({
        "EstadÃ­stico AÂ²": [A2],
        "p-value": [p]
    })
    return df


tbnormales = tabla_anderson_con_pvalue(resultado_normales)
tbnnormales = tabla_anderson_con_pvalue(resultado_no_normales)

# Nivel de significancia: 5%  â†’ 95% de confianza
alpha = 0.05

# -----------------------------
# 4. GrÃ¡ficas
# -----------------------------
def graficar_anderson(datos, titulo):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Histograma + densidad normal ajustada
    sns.histplot(datos, kde=True, stat="density", ax=axes[0],
                 color="skyblue", edgecolor="black")

    mu, sigma = np.mean(datos), np.std(datos)
    x = np.linspace(min(datos), max(datos), 100)
    y = stats.norm.pdf(x, mu, sigma)
    axes[0].plot(x, y, 'r-', lw=2, label=f"n({mu:.2f}, {sigma:.2f}Â²)")
    axes[0].set_title(f"Histograma - {titulo}")
    axes[0].legend()

    # QQ-plot
    stats.probplot(datos, dist="norm", plot=axes[1])
    axes[1].set_title(f"QQ Plot - {titulo}")

    plt.tight_layout()
    plt.show()


# -----------------------------
# 5. Resultados e interpretaciÃ³n al 95%
# -----------------------------
print("### Resultados - Datos normales (95% de confianza)")
display(tbnormales)

p_normales = tbnormales["p-value"][0]
if p_normales < alpha:
    print(f"ConclusiÃ³n (95%): p-value = {p_normales:.4f} < {alpha} â†’ Se rechaza Hâ‚€.")
    print("Los datos NO son normales al 95% de confianza.")
else:
    print(f"ConclusiÃ³n (95%): p-value = {p_normales:.4f} â‰¥ {alpha} â†’ No se rechaza Hâ‚€.")
    print("Los datos son compatibles con la normalidad al 95% de confianza.")

graficar_anderson(datos_normales, "Datos normales")


print("\n### Resultados - Datos no normales (95% de confianza)")
display(tbnnormales)

p_no_normales = tbnnormales["p-value"][0]
if p_no_normales < alpha:
    print(f"ConclusiÃ³n (95%): p-value = {p_no_normales:.4f} < {alpha} â†’ Se rechaza Hâ‚€.")
    print("Los datos NO son normales al 95% de confianza.")
else:
    print(f"ConclusiÃ³n (95%): p-value = {p_no_normales:.4f} â‰¥ {alpha} â†’ No se rechaza Hâ‚€.")
    print("Los datos son compatibles con la normalidad al 95% de confianza.")

graficar_anderson(datos_no_normales, "Datos no normales")

````

##6.8. Kolmogorovâ€“Smirnov (KS) Test (dos muestras)

Esta prueba compara las **distribuciones completas** de dos muestras continuas (no solo su media o mediana).

---
Â¿En quÃ© consiste?

La prueba KS de 2 muestras compara las **funciones de distribuciÃ³n empÃ­ricas (ECDF)** de dos grupos y mide quÃ© tanto se separan entre sÃ­.

- **EstadÃ­stico D:**  
  Es la **mÃ¡xima distancia vertical** entre las ECDF de las dos muestras.

- **p-value (valor p):**  
  Es la probabilidad de observar una diferencia **igual o mayor** que D si ambas muestras provinieran realmente de la **misma distribuciÃ³n**.

---
HipÃ³tesis

- **Hâ‚€:** ambas muestras provienen de la **misma distribuciÃ³n**.  
- **Hâ‚:** **al menos una** de las muestras proviene de una distribuciÃ³n diferente (prueba bilateral).

---
EstadÃ­stico D de Kolmogorovâ€“Smirnov

- Sea $F_{n_1}(x)$ la ECDF de la primera muestra de tamaÃ±o $n_1$.  
- Sea $G_{n_2}(x)$ la ECDF de la segunda muestra de tamaÃ±o $n_2$.

Entonces:

$D = \max_{x} \big| F_{n_1}(x) - G_{n_2}(x) \big|$

Es decir, buscamos el punto donde las dos curvas acumuladas estÃ¡n **mÃ¡s separadas verticalmente**.

---
FÃ³rmula (teÃ³rica) para el valor p

El **valor p** se calcula a partir de la estadÃ­stica \(D\) y los tamaÃ±os de muestra.  
Una forma clÃ¡sica de expresarlo es:

$\lambda = D \sqrt{\frac{n_1 n_2}{n_1 + n_2}}$

$p = 2 \sum_{j=1}^{\infty} (-1)^{j-1} e^{-2 j^2 \lambda^2}$

En la prÃ¡ctica, las librerÃ­as estadÃ­sticas (como SciPy) **no usan la suma infinita directamente**, sino aproximaciones numÃ©ricas equivalentes para obtener el p-value.

---

Criterio de decisiÃ³n
Por ejemplo al 95% de confianza:

- Si  $\text{p-value} < 0.05$
  â†’ **Se rechaza Hâ‚€.**  
  Hay evidencia estadÃ­stica de que las dos muestras **no** siguen la misma distribuciÃ³n.

- Si $\text{p-value} \geq 0.05$
  â†’ **No se rechaza Hâ‚€.**  
  No hay evidencia suficiente para afirmar que las distribuciones sean diferentes.

---
Ventajas y consideraciones

-Compara la **forma completa** de las distribuciones.  
- Supone datos **continuos**; si hay muchos empates (datos discretos), el p-value puede ser **conservador**.  
- Las muestras deben ser **independientes** y se compara una sola variable cuantitativa.

##6.9. (Opcional) Normalidad Multivariada (Henzeâ€“Zirkler)

La **normalidad multivariada** es un supuesto clave en muchos mÃ©todos estadÃ­sticos multivariados,  
como **MANOVA**, **regresiÃ³n mÃºltiple**, **anÃ¡lisis de componentes principales (PCA)** o **anÃ¡lisis discriminante**.

---

### ðŸ”¹ HipÃ³tesis

- **Hâ‚€:** los datos provienen de una distribuciÃ³n normal multivariada.  
- **Hâ‚:** los datos **no** siguen una distribuciÃ³n normal multivariada.

---

###  ðŸ”¹Etapas del anÃ¡lisis

1. **Normalidad univariada (por variable):**  
   Se evalÃºa con el **test de Shapiroâ€“Wilk**, que contrasta la forma de cada variable individualmente.  
   Si $p > 0.05$, se considera que la variable es aproximadamente normal.

2. **Normalidad multivariada (global):**  
   Se evalÃºa con el **test de Henzeâ€“Zirkler (HZ)**, mediante la funciÃ³n  
   `pingouin.multivariate_normality(df, alpha=0.05)`.

---

### ðŸ”¹ InterpretaciÃ³n

- Si **p > 0.05**, no se rechaza Hâ‚€ â†’ los datos **siguen una distribuciÃ³n normal multivariada**.  
- Si **p < 0.05**, se rechaza Hâ‚€ â†’ los datos **no siguen una distribuciÃ³n normal multivariada**.

---

### ðŸ”¹ FÃ³rmula del estadÃ­stico Henzeâ€“Zirkler

El estadÃ­stico $HZ$ se basa en una medida de distancia entre la funciÃ³n de densidad multivariada empÃ­rica  
y la teÃ³rica bajo normalidad, ponderada por la matriz de covarianza.

Aunque su cÃ¡lculo es complejo, en la prÃ¡ctica:
- Valores **pequeÃ±os de HZ** y **p grandes (> 0.05)** â†’ indican normalidad.  
- Valores **altos de HZ** y **p pequeÃ±os (< 0.05)** â†’ indican no normalidad.

---

### ðŸ“ˆ VisualizaciÃ³n

El **pairplot** o matriz de dispersiÃ³n permite inspeccionar visualmente las relaciones entre variables:  
- Si las nubes de puntos son **elÃ­pticas y simÃ©tricas**, se sugiere normalidad multivariada.  
- Si son **asimÃ©tricas o con colas largas**, podrÃ­a existir no normalidad.

---

  *Communications in Statistics - Theory and Methods*, 19(10), 3595â€“3617.  
- Shapiro, S. S., & Wilk, M. B. (1965). *An analysis of variance test for normality.* *Biometrika*, 52(3â€“4), 591â€“611.

````python
# --- Prueba de normalidad multivariada (Henzeâ€“Zirkler) ---

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro
from pingouin import multivariate_normality

# Reproducibilidad
np.random.seed(20)

# --- 1ï¸âƒ£ Generar dataset multivariado ---
# 7 variables, 50 observaciones
df = pd.DataFrame({
    f"x{i+1}": np.random.normal(size=50) for i in range(7)
})

# --- 2ï¸âƒ£ Evaluar normalidad univariada (Shapiroâ€“Wilk por variable) ---
print("=== Prueba de normalidad univariada (Shapiroâ€“Wilk) ===\n")
for col in df.columns:
    stat, p = shapiro(df[col])
    print(f"{col}: estadÃ­stico = {stat:.3f}, p = {p:.4f}")
    if p > 0.05:
        print(" â†’ No se rechaza Hâ‚€: distribuciÃ³n normal.\n")
    else:
        print(" â†’ Se rechaza Hâ‚€: no normal.\n")

# --- 3ï¸âƒ£ VisualizaciÃ³n univariada ---
sns.pairplot(df, diag_kind="kde")
plt.suptitle("DistribuciÃ³n de las variables univariadas", y=1.02)
plt.show()

# --- 4ï¸âƒ£ Prueba de normalidad multivariada (Henzeâ€“Zirkler) ---
hz_stat, p_value, normal = multivariate_normality(df, alpha=0.05)
print("\n=== Prueba Henzeâ€“Zirkler ===")
print(f"EstadÃ­stico HZ = {hz_stat:.3f}")
print(f"Valor p = {p_value:.4f}")
print(f"ConclusiÃ³n: {'Datos normales multivariados' if normal else 'Datos no normales multivariados'}")
````

Dado que el valor p de la prueba no es menor que nuestro valor alfa especificado de .05, no rechazamos la hipÃ³tesis nula. Se puede suponer que el conjunto de datos sigue una distribuciÃ³n normal multivariante.

### ðŸš« Ejemplo: Datos que no cumplen normalidad multivariada

En este ejemplo se genera un conjunto de datos con **siete variables de diferentes distribuciones**  
(normales, lognormales, exponenciales, Ï‡Â², beta y uniformes).  
Esto rompe explÃ­citamente el supuesto de **multinormalidad**.

---

### ðŸ”¹ HipÃ³tesis

- **Hâ‚€:** los datos provienen de una distribuciÃ³n normal multivariada.  
- **Hâ‚:** los datos **no** provienen de una distribuciÃ³n normal multivariada.

---

### ðŸ”¹ Resultados esperados

- La mayorÃ­a de las pruebas **Shapiroâ€“Wilk** por variable muestran **p < 0.05**,  
  indicando que las variables no son normales individualmente.  
- La **prueba Henzeâ€“Zirkler** tambiÃ©n arroja un **p < 0.05**,  
  por lo tanto se **rechaza Hâ‚€**: los datos **no son normales multivariados**.

### ðŸ”¹ InterpretaciÃ³n visual

El grÃ¡fico de pares (`pairplot`) muestra:

### ðŸ”¹ ConclusiÃ³n

> Los datos analizados **no cumplen el supuesto de normalidad multivariada**,  
> por lo tanto **no son adecuados para tÃ©cnicas como MANOVA, PCA o discriminante lineal**,  
> a menos que se realicen **transformaciones** (ej., logarÃ­tmica)

````python
# --- Prueba de normalidad multivariada (caso no normal) ---

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro
from pingouin import multivariate_normality

# Reproducibilidad
np.random.seed(20)

# --- 1ï¸âƒ£ Generar dataset multivariado no normal ---
# Mezcla de distribuciones: algunas normales, otras lognormales o sesgadas
df_no_normal = pd.DataFrame({
    'x1': np.random.normal(size=50),
    'x2': np.random.lognormal(mean=0, sigma=1, size=50),
    'x3': np.random.exponential(scale=1, size=50),
    'x4': np.random.chisquare(df=3, size=50),
    'x5': np.random.beta(a=2, b=5, size=50) * 10,
    'x6': np.random.uniform(low=-2, high=2, size=50),
    'x7': np.random.normal(loc=5, scale=2, size=50)
})

# --- 2ï¸âƒ£ Prueba univariada de Shapiroâ€“Wilk ---
print("=== Prueba de normalidad univariada (Shapiroâ€“Wilk) ===\n")
for col in df_no_normal.columns:
    stat, p = shapiro(df_no_normal[col])
    print(f"{col}: estadÃ­stico = {stat:.3f}, p = {p:.4f}")
    if p > 0.05:
        print(" â†’ No se rechaza Hâ‚€: distribuciÃ³n normal.\n")
    else:
        print(" â†’ Se rechaza Hâ‚€: no normal.\n")

# --- 3ï¸âƒ£ VisualizaciÃ³n ---
sns.pairplot(df_no_normal, diag_kind="kde")
plt.suptitle("DistribuciÃ³n de variables â€” Caso no normal multivariado", y=1.02)
plt.show()

# --- 4ï¸âƒ£ Prueba de normalidad multivariada (Henzeâ€“Zirkler) ---
hz_stat, p_value, normal = multivariate_normality(df_no_normal, alpha=0.05)
print("\n=== Prueba Henzeâ€“Zirkler ===")
print(f"EstadÃ­stico HZ = {hz_stat:.3f}")
print(f"Valor p = {p_value:.4f}")
print(f"ConclusiÃ³n: {'Datos normales multivariados' if normal else 'Datos no normales multivariados'}")
````

### ðŸ“ˆ GrÃ¡ficos Qâ€“Q (Quantileâ€“Quantile Plot)

Un **Qâ€“Q plot** permite verificar visualmente si un conjunto de datos sigue una **distribuciÃ³n normal**.

- El **eje X** muestra los *cuantiles teÃ³ricos* (valores que tendrÃ­a la muestra si fuera normal).  
- El **eje Y** muestra los *cuantiles observados* (valores reales del dataset).  
- Si los puntos se alinean cerca de la **lÃ­nea de 45Â°**, los datos se consideran aproximadamente normales.

ðŸ”¹ En el ejemplo:
- Los datos generados con `np.random.normal()` siguen la lÃ­nea â†’ **normalidad confirmada**.  
- Los datos generados con `np.random.uniform()` se desvÃ­an de la lÃ­nea â†’ **no normalidad**.


````python
# --- Ejemplo 1: Qâ€“Q plot normal vs uniforme ---

import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

np.random.seed(42)

# Datos normales y uniformes
data_normal = np.random.normal(0, 1, 100)
data_uniform = np.random.uniform(0, 1, 100)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Normal
sm.qqplot(data_normal, line='45', ax=axes[0])
axes[0].set_title("DistribuciÃ³n normal")

# Uniforme
sm.qqplot(data_uniform, line='45', ax=axes[1])
axes[1].set_title("DistribuciÃ³n uniforme (no normal)")

plt.tight_layout()
plt.show()
````

##6.10. (Opcional) *Outliers* (valores atÃ­picos)

Los **outliers** son observaciones que se desvÃ­an significativamente del resto de los datos.  
Pueden representar **errores, excepciones o casos de interÃ©s** dependiendo del contexto.

---

### ðŸ”¹ Tipos de outliers segÃºn su significado

| Tipo | Ejemplo | AcciÃ³n recomendada |
|------|----------|--------------------|
| **Error** | Edad de 160 aÃ±os en humanos | Corregir o eliminar |
| **LÃ­mite vÃ¡lido** | Temperatura muy alta pero real | Mantener o ajustar |
| **Caso especial o interÃ©s** | Fraude detectado por su patrÃ³n anÃ³malo | Analizar separadamente |

---

### ðŸ”¹Â¿Por quÃ© detectarlos?

Los outliers pueden:
- **Distorsionar** la media y la varianza.
- **Sesgar** los modelos predictivos (p. ej. regresiÃ³n lineal, PCA, ensambles).
- **Ocultar relaciones** importantes en los datos.

Por ello, antes de aplicar pruebas estadÃ­sticas o modelos de machine learning,  
es buena prÃ¡ctica **identificarlos, analizarlos y decidir su tratamiento.**

###ðŸ§® 5ï¸âƒ£ MÃ©todos formales de detecciÃ³n

AdemÃ¡s de inspecciones visuales, existen pruebas estadÃ­sticas para detectar outliers de forma objetiva:

| Prueba             | AplicaciÃ³n                      | HipÃ³tesis nula (Hâ‚€)                         | InterpretaciÃ³n                                 |
| ------------------ | ------------------------------- | ------------------------------------------- | ---------------------------------------------- |
| **Dixon (Q-test)** | Muestras pequeÃ±as (n â‰¤ 30)      | No hay valores atÃ­picos                     | Si p < 0.05 â†’ se rechaza Hâ‚€, hay outlier       |
| **Grubbs**         | Una muestra, un posible outlier | No hay outliers                             | Si p < 0.05 â†’ se rechaza Hâ‚€, hay outlier       |
| **IQR (Tukey)**    | General, robusta                | No hay valores fuera del rango intercuartil | Si valor < Q1â€“1.5Ã—IQR o > Q3+1.5Ã—IQR â†’ outlier |


##Pruebas Dixon y Grubbs para detecciÃ³n de outliers en ejemoplo de edades

### ðŸ”¹ InterpretaciÃ³n prÃ¡ctica

- **Grubbs:** compara la desviaciÃ³n del valor extremo con respecto a la media.  
- **Dixon:** compara la distancia entre el valor mÃ¡s extremo y su vecino inmediato  
  con respecto al rango total de la muestra.

---

### ðŸ”¹ Resultado esperado en este caso

Para el conjunto de datos:

`[22, 22, 23, ..., 34, 80]`

- El valor **80** se identifica como **outlier** por ambas pruebas.  
- En el test de Dixon:  
  $Q_{calc} = 0.53 > Q_{crit} = 0.33$ â†’ se **rechaza Hâ‚€** (p < 0.05).  
- En el test de Grubbs:  
  $p < 0.05$ â†’ tambiÃ©n se **rechaza Hâ‚€**, confirmando el outlier.

---

ðŸ”¹ **Referencias**
- S. P. Verma & A. Quiroz-Ruiz (2006). *Critical values for six Dixon tests for outliers in normal samples up to size 100.*  
- Grubbs, F. E. (1950). *Sample criteria for testing outlying observations.* *Annals of Mathematical Statistics*, 21(1), 27â€“58.

````python
# --- DetecciÃ³n de outliers con Grubbs y Dixon (versiÃ³n sin librerÃ­as externas) ---

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Datos de ejemplo
edades = np.array([22,22,23,23,23,23,26,27,27,28,30,30,30,30,31,32,33,34,80])
alpha = 0.05
n = len(edades)

print(f"=== AnÃ¡lisis de outliers (n={n}) ===\n")
print(f"Datos: {edades}\n")

# --- 1ï¸âƒ£ Prueba de Grubbs (manual con scipy) ---
mean_y = np.mean(edades)
std_y = np.std(edades, ddof=1)
abs_diffs = np.abs(edades - mean_y)
max_deviation = np.max(abs_diffs)
outlier_candidate = edades[np.argmax(abs_diffs)]
G_cal = max_deviation / std_y

# Valor crÃ­tico de Grubbs (bilateral)
t_crit = stats.t.ppf(1 - alpha / (2 * n), n - 2)
G_crit = ((n - 1) / np.sqrt(n)) * np.sqrt(t_crit**2 / (n - 2 + t_crit**2))

print("ðŸ”¹ Prueba de Grubbs (dos colas):")
print(f"EstadÃ­stico G = {G_cal:.3f}, GcrÃ­tico = {G_crit:.3f}")
if G_cal > G_crit:
    print(f" â†’ Se rechaza Hâ‚€: {outlier_candidate} es un outlier (p < 0.05).\n")
    outlier_grubbs = outlier_candidate
else:
    print(" â†’ No se rechaza Hâ‚€: no hay outlier segÃºn Grubbs.\n")
    outlier_grubbs = None

# --- 2ï¸âƒ£ Prueba de Dixon (Q-test) ---
def dixon_test(data, alpha=0.05, side='max'):
    data = np.sort(data)
    n = len(data)
    if n < 3 or n > 30:
        raise ValueError("El test de Dixon solo es vÃ¡lido para 3 â‰¤ n â‰¤ 30.")

    Q_table = {  # Î±=0.05
        3: 0.941, 4: 0.765, 5: 0.642, 6: 0.560, 7: 0.507, 8: 0.468,
        9: 0.437, 10: 0.412, 11: 0.392, 12: 0.376, 13: 0.361, 14: 0.349,
        15: 0.338, 16: 0.329, 17: 0.320, 18: 0.313, 19: 0.306, 20: 0.300,
        21: 0.294, 22: 0.289, 23: 0.284, 24: 0.279, 25: 0.275, 26: 0.271,
        27: 0.268, 28: 0.264, 29: 0.261, 30: 0.258
    }

    if side == 'max':
        Q_calc = (data[-1] - data[-2]) / (data[-1] - data[0])
        candidate = data[-1]
    else:
        Q_calc = (data[1] - data[0]) / (data[-1] - data[0])
        candidate = data[0]

    Qcrit = Q_table.get(n, None)
    return Q_calc, Qcrit, candidate

Q_calc, Qcrit, candidate = dixon_test(edades, alpha=alpha, side='max')
print("ðŸ”¹ Prueba de Dixon (valor mÃ¡ximo):")
print(f"EstadÃ­stico Q = {Q_calc:.3f}, QcrÃ­tico = {Qcrit:.3f}")
if Q_calc > Qcrit:
    print(f" â†’ Se rechaza Hâ‚€: {candidate} es un outlier segÃºn Dixon (p < 0.05).")
    outlier_dixon = candidate
else:
    print(" â†’ No se rechaza Hâ‚€: no hay evidencia de outlier segÃºn Dixon.")
    outlier_dixon = None

# --- 3ï¸âƒ£ VisualizaciÃ³n ---
sns.set(style="whitegrid")
plt.figure(figsize=(7,4))
sns.boxplot(x=edades, color="lightblue", width=0.5)

if outlier_grubbs:
    plt.scatter(outlier_grubbs, 0, color='red', s=120, label=f"Grubbs ({outlier_grubbs})", zorder=3)
if outlier_dixon and outlier_dixon != outlier_grubbs:
    plt.scatter(outlier_dixon, 0, color='orange', s=120, label=f"Dixon ({outlier_dixon})", zorder=3)

plt.title("Outliers detectados por Grubbs y Dixon")
plt.xlabel("Valores de la variable (Edad)")
plt.legend()
plt.show()

````

- `weight` estÃ¡ sesgada a la derecha (hay una mayor concentraciÃ³n entre 60 y 90 kg), por lo que se parece a una log-normal.

### 1. Pruebas de normalidad (Shapiroâ€“Wilk y Kolmogorovâ€“Smirnov) para todas las variables cuantitativas

````python
resultados_pruebas = []

for var in vars_cuantitativas:
    data = df[var].dropna()

    #Shapiro-Wilk
    shapiro_stat, shapiro_p = stats.shapiro(data.sample(5000))

    #Kolmogorov-Smirnov (comparaciÃ³n con una normal ajustada)
    mu, sigma = data.mean(), data.std(ddof=0)
    kolmogorov_stat, kolmogorov_p = stats.kstest((data - mu) / sigma, 'norm')

    #Interpretaciones
    interp_shapiro = 'Normal (pâ‰¥0.05)' if shapiro_p >= 0.05 else 'No normal (p<0.05)'
    interp_kolmogorov = 'Normal (pâ‰¥0.05)' if kolmogorov_p >= 0.05 else 'No normal (p<0.05)'

    resultados_pruebas.append([shapiro_stat, shapiro_p, interp_shapiro, kolmogorov_stat, kolmogorov_p, interp_kolmogorov])

pruebas_normalidad = pd.DataFrame(data=resultados_pruebas, index=vars_cuantitativas,
                                  columns=['Shapiro-Wilk', 'p-value Shapiro', 'InterpretaciÃ³n Shapiro',
                                           'Kolmogorov-Smirnov', 'p-value Kolmogorov', 'InterpretaciÃ³n Kolmogorov'])
pruebas_normalidad
````

### 2. Ajuste visual y comparaciÃ³n con distribuciones teÃ³ricas

````python
# Diccionario que indica, para cada variable cuantitativa,
# quÃ© distribuciones teÃ³ricas queremos ajustar y comparar.
distribuciones = {
    'age':    ['normal'],
    'height': ['normal', 't-student'],
    'weight': ['normal', 'log-normal'],
    'ap_hi':  ['normal', 'pareto'],
    'ap_lo':  ['normal', 'pareto']
}

# Recorremos todas las variables cuantitativas de interÃ©s
for var in vars_cuantitativas:
    # Tomamos los datos de la variable y eliminamos NA
    data = df[var].dropna()

    # Figura base: histograma + KDE para ver la distribuciÃ³n empÃ­rica
    plt.figure(figsize=(10,5))
    sns.histplot(
        data,
        bins=70,
        kde=True,          # curva suavizada (estimaciÃ³n de densidad)
        stat='density',    # para que el Ã¡rea se interprete como densidad
        color='#8882d9',
        alpha=0.25
    )

    # Definimos un rango de valores en el eje X para evaluar las distribuciones teÃ³ricas
    x = np.linspace(data.min(), data.max(), 200)

    # Para cada distribuciÃ³n teÃ³rica asociada a esta variable...
    for dist in distribuciones[var]:

        # 1) Ajuste y grÃ¡fica de la distribuciÃ³n normal
        if dist == 'normal':
            # Ajusta media (mu) y desviaciÃ³n estÃ¡ndar (sigma) por mÃ¡xima verosimilitud
            mu, sigma = stats.norm.fit(data)
            # Grafica la densidad teÃ³rica normal ajustada sobre el histograma
            plt.plot(x, stats.norm.pdf(x, mu, sigma),
                     label='Normal', color='#ba206d')

        # 2) Ajuste y grÃ¡fica de la t de Student
        elif dist == 't-student':
            # Ajusta grados de libertad, localizaciÃ³n y escala
            df_t, loc_t, scale_t = stats.t.fit(data)
            plt.plot(x, stats.t.pdf(x, df_t, loc_t, scale_t),
                     label='t-Student', color="#0db2ac")

        # 3) Ajuste y grÃ¡fica de la log-normal
        elif dist == 'log-normal':
            # Ajuste de lognormal, fijando loc=0 (datos estrictamente positivos)
            shape, loc, scale = stats.lognorm.fit(data, floc=0)
            plt.plot(x, stats.lognorm.pdf(x, shape, loc, scale),
                     label='Log-normal', color="#0db2ac")

        # 4) Ajuste y grÃ¡fica de Pareto
        elif dist == 'pareto':
            # Filtra valores > 1 para evitar problemas en el ajuste de Pareto
            b, loc, scale = stats.pareto.fit(data[data > 1])
            plt.plot(x, stats.pareto.pdf(x, b, loc, scale),
                     label='Pareto', color='#0db2ac')

    # TÃ­tulo y leyenda para identificar la variable y las distribuciones comparadas
    plt.title(f'ComparaciÃ³n de distribuciones teÃ³ricas - {var}')
    plt.legend()
    plt.show()
````

## **Parte 3. Pruebas de HipÃ³tesis**

### 1. EvaluaciÃ³n de supuestos previos

##### Verificar normalidad (por grupo de cardio)

````python
resultados_normalidad = []

for var in vars_cuantitativas:
    grupo0 = df[df['cardio'] == 0][var].dropna()
    grupo1 = df[df['cardio'] == 1][var].dropna()

    #Shapiro-Wilk para ambos grupos
    stat0, p0 = stats.shapiro(grupo0.sample(5000))
    stat1, p1 = stats.shapiro(grupo1.sample(5000))

    resultados_normalidad.append([p0, 'Normal' if p0 >= 0.05 else 'No normal',
                                  p1, 'Normal' if p1 >= 0.05 else 'No normal'])

normalidad = pd.DataFrame(data=resultados_normalidad, index=vars_cuantitativas,
                          columns=['p-value grupo 0', 'InterpretaciÃ³n grupo 0', 'p-value grupo 1', 'InterpretaciÃ³n grupo 1'])
normalidad
````

##### Evaluar homogeneidad de varianzas con pruebas de Levene y Bartlett

````python
resultados_varianzas = []

for var in vars_cuantitativas:
    grupo0 = df[df['cardio'] == 0][var].dropna()
    grupo1 = df[df['cardio'] == 1][var].dropna()

    #Pruebas de homogeneidad
    stat_lev, p_lev = stats.levene(grupo0, grupo1)
    stat_bart, p_bart = stats.bartlett(grupo0, grupo1)

    resultados_varianzas.append([stat_lev, p_lev, 'HomogÃ©neas' if p_lev >= 0.05 else 'No homogÃ©neas',
                                 stat_bart, p_bart, 'HomogÃ©neas' if p_bart >= 0.05 else 'No homogÃ©neas'])

homogeneidad_varianzas = pd.DataFrame(data=resultados_varianzas, index=vars_cuantitativas,
                                      columns=['EstadÃ­stico Levene', 'p-value Levene', 'InterpretaciÃ³n Levene',
                                               'EstadÃ­stico Bartlett', 'p-value Bartlett', 'InterpretaciÃ³n Bartlett'])
homogeneidad_varianzas
````

### 2. SelecciÃ³n de la prueba adecuada

No se cuemplen los supuestos (ni normalidad ni homocedasticidad -igualdad de varianzas), por lo que se aplicarÃ¡ la prueba **no paramÃ©trica Mannâ€“Whitney U**.

````python
resultados_mann_whitney = []

for var in vars_cuantitativas:
    grupo0 = df[df['cardio'] == 0][var].dropna()
    grupo1 = df[df['cardio'] == 1][var].dropna()

    stat, p = stats.mannwhitneyu(grupo0, grupo1, alternative='two-sided')

    interpretacion = 'Diferencia significativa' if p < 0.05 else 'Sin diferencia significativa'

    resultados_mann_whitney.append([stat, p, interpretacion])

mann_whitney_u = pd.DataFrame(data=resultados_mann_whitney, index=vars_cuantitativas, columns=['Mann-Whitney U', 'p-value', 'InterpretaciÃ³n'])
mann_whitney_u
````

Todas las variables presentan diferencias estadÃ­sticamente significativas entre los grupos con y sin enfermedad cardiovascular.

Las personas con enfermedad cardiovascular tienden a tener mayor edad, peso y presiÃ³n arterial que aquellas sin la enfermedad, mientras que tambiÃ©n se observan diferencias menores pero significativas en la altura.

En conjunto, estos resultados sugieren que las variables evaluadas estÃ¡n asociadas con la presencia de enfermedad cardiovascular.

# Parte 2. Distribuciones de Probabilidad

Usaremos la prueba de **Shapiro-Wilk** para validar si las variables son normales usando pruebas estadÃ­sticas.

````python
# Pruebas de Shapiro-Wilk para las variables cuantitativas
print("Resultados de la Prueba de Shapiro-Wilk:")

# Prueba para total_bill
stat_bill, p_bill = stats.shapiro(data['total_bill'])
print(f"Total Bill: p-valor = {p_bill}")

# Prueba para tip
stat_tip, p_tip = stats.shapiro(data['tip'])
print(f"Tip: p-valor = {p_tip}")

# Prueba para size
stat_size, p_size = stats.shapiro(data['size'])
print(f"Size: p-valor = {p_size}")
````

# Parte 3. Pruebas de HipÃ³tesis

**HipÃ³tesis:** Â¿El monto de la propina (`tip`) es diferente entre clientes fumadores (`smoker`) y no fumadores?

**EvaluaciÃ³n de supuestos previos**

* **Normalidad:** Ya sabemos que la variable `tip` no es normal.

* **Homogeneidad de varianzas:** Verificamos si la dispersiÃ³n de las propinas es igual en ambos grupos (fumadores y no fumadores). Usando la **Prueba de Levene** para ver si las varianzas eran iguales (homogeneidad).

````python
# Separar las propinas por grupo
tips_fumadores = data[data['smoker'] == 'Yes']['tip']
tips_no_fumadores = data[data['smoker'] == 'No']['tip']

# Prueba de Levene
stat_levene, p_levene = stats.levene(tips_fumadores, tips_no_fumadores)
print(f"Prueba de Levene: p-valor = {p_levene}")
````

**Resultado de Levene:**

`p_levene` es > 0.05, las varianzas son homogÃ©neas (iguales).

**SelecciÃ³n de Prueba**

* El supuesto de Normalidad **NO** se cumpliÃ³

* El supuesto de Varianzas **SÃ** se cumpliÃ³

Como uno de los supuestos (Normalidad) fallÃ³, asi que se descarta la prueba paramÃ©trica (**T-Student**) y se elige la prueba no paramÃ©trica: **U de Mann-Whitney**.

````python
# Aplicar la Prueba U de Mann-Whitney
stat_mann, p_mann = stats.mannwhitneyu(tips_fumadores, tips_no_fumadores)

print(f"Prueba U de Mann-Whitney: p-valor = {p_mann}")
````
